```python
import math
from IPython.display import clear_output

# =====================================================================
# 1. TRAINING PHASE (Engine sirf ek baar chalega - EKDUM SAME)
# =====================================================================
X_train = 2.0
Y_train = 1.0
m = 1.5
c = -2.0
learning_rate = 0.1
purana_loss = 999999.0

while True:
    Z = (m * X_train) + c
    P = 1 / (1 + math.exp(-Z))
    loss = - (Y_train * math.log(P) + (1 - Y_train) * math.log(1 - P))
    gradient_c = P - Y_train
    gradient_m = X_train * (P - Y_train)
    
    if abs(purana_loss - loss) < 0.000001:
        break
        
    m = m - (learning_rate * gradient_m)
    c = c - (learning_rate * gradient_c)
    purana_loss = loss

real_m = m
real_c = c


# =====================================================================
# 2. UPDATED LIVE TERMINAL (Alphabet safe with persistent message)
# =====================================================================
default_X = 2.5
error_msg = ""  # Yeh variable galat input wale message ko yaad rakhega

while True:
    # Screen saaf hogi
    clear_output(wait=True)
    
    print("✅ ENGINE TRAINED SUCCESSFULLY!")
    print(f"Locked Parameters -> Real m: {real_m:.4f} | Real c: {real_c:.4f}")
    print("=" * 60)
    print("👉 LIVE SALES PREDICTION ENGINE (Type 'exit' to stop) 👈")
    print("=" * 60)
    
    # Live Calculation using locked weights
    Z_naya = (real_m * default_X) + real_c
    P_naya = 1 / (1 + math.exp(-Z_naya))
    percentage = P_naya * 100
    
    print(f"\nLast Valid Input Processed (X): {default_X}")
    print(f"Formula Output       -> Z = ({real_m:.4f} * {default_X}) + ({real_c:.4f}) = {Z_naya:.4f}")
    print(f"Khareedne ka chance  : {percentage:.2f}%")
    
    if P_naya >= 0.5:
        print("🟢 FINAL ANSWER      : HAAN (Gold designs dikhao!)")
    else:
        print("🔴 FINAL ANSWER      : NAA (Chances kam hain)")
    print("-" * 60)
    
    # AGAR PURANE CHAKKAR ME ERROR AAYA THA, TO WOH YAHAN SCREEN PAR FIX REHEGA
    if error_msg:
        print(error_msg)
        print("-" * 60)
    
    # User se input lena
    user_input = input("Bhai, naya Score (X) dalo aur Enter dabao: ").strip()
    
    # Exit condition
    if user_input.lower() == 'exit':
        print("\nPrediction engine band ho gaya. Shukriya bhai!")
        break
        
    # Input validation block
    try:
        default_X = float(user_input)
        error_msg = ""  # Agar number sahi hai, toh error message ko saaf kar do
    except ValueError:
        # Agar alphabet dala, toh message is variable mein save ho jayega aur agle loop mein bhi dikhta rahega
        error_msg = f"❌ Kripya sahi number enter karein! ('{user_input}' ek number nahi hai)"
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# 1. DATA INPUT (Ab hum 2 examples de rahe hain taaki sklearn seekh sake)
# X_train: Pehle customer ka score 2.0, doosre ka 0.5
X_train = np.array([[2.0], [0.5]]) 

# Y_train: Pehle ne khareeda (1), doosre ne nahi khareeda (0)
Y_train = np.array([1, 0])     

# 2. EK LINE MEIN TRAINING (Ab error nahi aayega!)
model = LogisticRegression().fit(X_train, Y_train)

print("✅ MODEL TRAINED AUTOMATICALLY BY SKLEARN!")
print(f"Auto-Calculated -> m: {model.coef_[0][0]:.4f} | c: {model.intercept_[0]:.4f}")
print("=" * 60)

# 3. LIVE PREDICTION ENGINE (Aap bas X doge, yeh automatic answer dega)
while True:
    clear_output(wait=True)
    inp = input("\nBhai, bas naya Score (X) dalo [ya 'exit']: ").strip()
    if inp.lower() == 'exit': break
    
    try:
        naya_X = float(inp)
        
        # Ek line mein direct probability (Chance)
        probability = model.predict_proba([[naya_X]])[0][1] * 100
        
        # Ek line mein direct prediction (0 ya 1)
        prediction = model.predict([[naya_X]])[0]
        
        print(f"-> Khareedne ka chance: {probability:.2f}%")
        print("🟢 FINAL ANSWER: HAAN" if prediction == 1 else "🔴 FINAL ANSWER: NAA")
        
    except ValueError:
        print("❌ Kripya sahi number enter karein!")
```
