```python
# =====================================================================
# STEP 1: DATA SETUP (Karigaron ka data aur unke pieces ka weight)
# =====================================================================
# Hamare teeno karigaron ke 3-3 pieces ka data lists me daal diya
abdullah = [10, 11, 9]
badshah  = [12, 13, 11]
vishal   = [8, 9, 7]

# Saare pieces ko ek sath mila kar factory ka total data banaya (Total N = 9)
all_data = abdullah + badshah + vishal
N = len(all_data) # Kul pieces = 9
k = 3             # Kul groups (karigar) = 3

print("--- STEP 1: DATA LOADED ---")
print(f"Total Pieces in Factory: {N}")
print(f"Total Artisans (Groups): {k}\n")


# =====================================================================
# STEP 2: CALCULATE MEANS (Averages nikalna)
# =====================================================================
# Har ek karigar ka apna personal average (Group Mean)
mean_abdullah = sum(abdullah) / len(abdullah) # (10+11+9)/3 = 10.0
mean_badshah  = sum(badshah) / len(badshah)   # (12+13+11)/3 = 12.0
mean_vishal   = sum(vishal) / len(vishal)     # (8+9+7)/3 = 8.0

# Poori factory ka ek single center point average (Grand Mean)
grand_mean = sum(all_data) / N # 90 / 9 = 10.0

print("--- STEP 2: MEANS CALCULATED ---")
print(f"Abdullah Mean: {mean_abdullah}")
print(f"Badshah Mean: {mean_badshah}")
print(f"Vishal Mean: {mean_vishal}")
print(f"Factory Grand Mean: {grand_mean}\n")


# =====================================================================
# STEP 3: MSW (Karigar Se Hui Error / Within Groups Variance)
# =====================================================================
# Pehle har karigar ke pieces ka unhi ke average se minus karke square nikalenge
ssw_abdullah = sum([(x - mean_abdullah)**2 for x in abdullah]) # (10-10)^2 + (11-10)^2 + (9-10)^2 = 2
ssw_badshah  = sum([(x - mean_badshah)**2 for x in badshah])   # (12-12)^2 + (13-12)^2 + (11-12)^2 = 2
ssw_vishal   = sum([(x - mean_vishal)**2 for x in vishal])     # (8-8)^2 + (9-8)^2 + (7-8)^2 = 2

# Saare personal errors ko jod kar factory ka total error banaya
ssw = ssw_abdullah + ssw_badshah + ssw_vishal # 2 + 2 + 2 = 6

# Error karne ke chances (Degree of Freedom 2) = Total Pieces - Total Groups
df2 = N - k # 9 - 3 = 6

# Average Error Unit nikalne ke liye divide kiya (MSW)
msw = ssw / df2 # 6 / 6 = 1.0

print("--- STEP 3: WITHIN GROUPS (MSW) ---")
print(f"Sum of Squares Within (SSW - Total Personal Error): {ssw}")
print(f"Degrees of Freedom 2 (df2 - Error Chances): {df2}")
print(f"Mean Square Within (MSW - Kismat ka Normal Jhatka): {msw} Gram\n")


# =====================================================================
# STEP 4: MSB (Group Se Huee Error / Between Groups Variance)
# =====================================================================
# Har group ke average ka Grand Mean se farq nikaal kar square kiya
# Aur imaandari ke paimane ke liye unke peeche khade 3-3 pieces se MULTIPLY kiya!
n_pieces = 3 # Har group me kitne samples hain

ssb_abdullah = n_pieces * (mean_abdullah - grand_mean)**2 # 3 * (10 - 10)^2 = 0
ssb_badshah  = n_pieces * (mean_badshah - grand_mean)**2  # 3 * (12 - 10)^2 = 12
ssb_vishal   = n_pieces * (mean_vishal - grand_mean)**2   # 3 * (8 - 10)^2 = 12

# Saare group errors ko jod diya
ssb = ssb_abdullah + ssb_badshah + ssb_vishal # 0 + 12 + 12 = 24

# Group ka azaadi ka number (Degree of Freedom 1) = Total Groups - 1
df1 = k - 1 # 3 - 1 = 2

# Ek group ke level par average aapas ka farq nikalne ke liye divide kiya (MSB)
msb = ssb / df1 # 24 / 2 = 12.0

print("--- STEP 4: BETWEEN GROUPS (MSB) ---")
print(f"Sum of Squares Between (SSB - Total Group Error): {ssb}")
print(f"Degrees of Freedom 1 (df1 - Group Azaadi): {df1}")
print(f"Mean Square Between (MSB - Group Ka Aapas Ka Farq): {msb} Grams\n")


# =====================================================================
# STEP 5: F-SCORE & FINAL DECISION (Dono Errors Ki Takkar)
# =====================================================================
# Dono values ko aapas me bhida kar ratio (Guna) nikalenge
f_score = msb / msw # 12.0 / 1.0 = 12.0

print("--- STEP 5: FINAL F-SCORE EVALUATION ---")
print(f"Calculated F-Score: {f_score}")
print(f"Iska matlab: Group ka aapas ka farq normal kismat ke jhatke se {f_score} GUNA bada hai!")

# Critical check jo hamne baat ki thi (Scale Blindness Check)
if f_score == 1.0 and msw > 5.0:
    print("🚨 SYSTEM RED ALERT: F-Score 1 hai par MSW bohot bada hai! Poora factory system kharab hai!")
elif f_score > 4.0:
    print("❌ NULL HYPOTHESIS REJECTED: Karigaron ke kaam me asli lafda hai, ittefaq nahi hai!")
else:
    print("✅ NULL HYPOTHESIS WINS: Sab normal hai, kismat ka chota-mota variation hai.")
```

```python
# =====================================================================
# STEP 1: LIBRARY IMPORT
# =====================================================================
# SciPy library se f_oneway (ANOVA) ka tool bula liya
from scipy.stats import f_oneway

# =====================================================================
# STEP 2: DATA SETUP (Karigaron ka data)
# =====================================================================
abdullah = [10, 11, 9]
badshah  = [12, 13, 11]
vishal   = [8, 9, 7]

# =====================================================================
# STEP 3: 1-LINE ANOVA MAGIC 🚀
# =====================================================================
# Yeh single line parde ke peeche MSB, MSW sab calculate kar degi!
f_score, p_value = f_oneway(abdullah, badshah, vishal)

print("--- SCI-PY ANOVA RESULTS ---")
print(f"Calculated F-Score : {f_score:.2f}") # Output: 12.00
print(f"P-Value            : {p_value:.4f}") # Output: 0.0080

# =====================================================================
# STEP 4: AUTOMATED DECISION
# =====================================================================
# Alpha = 0.05 (Hamara 5% ka standard shakk ka paimana)
if p_value < 0.05:
    print("\n❌ NULL HYPOTHESIS REJECTED: Karigaron ke kaam me asli farq hai, koi ittefaq nahi hai!")
else:
    print("\n✅ NULL HYPOTHESIS WINS: Sab normal hai, kismat ka chota-mota variation hai.")
```
