```python
# Sone ki rings ka wazan (Weights in grams)
ring_weights = [5.4, 4.8, 6.2, 5.0, 5.5, 5.9, 4.5]

# Factory workers ki ginti (Daily Staff count)

# Sabhi weights ko jodo (Sum) aur unki ginti (Count) se divide karo
total_weight = sum(ring_weights)
number_of_rings = len(ring_weights)

mean_weight = total_weight / number_of_rings

print(f"Total Weight: {total_weight}g")
print(f"Total Rings: {number_of_rings}")
print(f"Mean Weight of a Ring: {mean_weight:.2f}g")
```

```python
staff_count = [12, 15, 14, 12, 15, 13]
import statistics

avg_staff = statistics.mean(staff_count)
print(f"Average Daily Staff: {avg_staff}")
```

```python
# Ek aisi list jisme ek outlier hai (80g ki ring galti se likh di gayi)
faulty_data = [5, 4, 6, 5, 5, 80] 
import statistics
# Iska mean nikal kar dekhiye
wrong_mean = statistics.mean(faulty_data)
print(f"Mean with Outlier: {wrong_mean}g")

# Logic: Agar mean normal values (4-6g) se bahut bada aa raha hai,
# iska matlab data mein outlier hai!
```
