```python
# ====================================================================================================
# Subject: Central Limit Theorem & Manual Calculus Probability Core
# ====================================================================================================

import math

# Step 0: Given Parameters Definition
mu = 221.55       # Asli Population Mean (AED)
sigma = 229.73    # Asli Population Standard Deviation (σ)
n = 40            # Sample Size / Invoices ka jhund
target = 250.0    # Target Group Mean jiska chance nikalna hai

print("============== STEP 1: PARAMETERS IDENTIFICATION ==============")
print(f"Population Mean (mu) : {mu} AED")
print(f"Population SD (sigma): {sigma} AED")
print(f"Sample Size (n)      : {n} bills")
print(f"Target Group Mean    : {target} AED\n")


# Step 1: Calculate Standard Error (SE) Engine Manual Math
# SE = sigma / sqrt(n)
sqrt_n = n ** 0.5
se = sigma / sqrt_n

print("============== STEP 2: STANDARD ERROR ENGINE ==============")
print(f"Square Root of n (sqrt(n))     : {sqrt_n:.5f}")
print(f"Fixed Standard Error Scale (SE): {se:.5f} AED\n")


# Step 2: Calculate Z-Score (Steps Counter Engine)
# Z = (Target - Mean) / SE
z_score = (target - mu) / se

print("============== STEP 3: Z-SCORE (STEPS COUNTER) ==============")
print(f"Actual Deviation (Target - mu) : {(target - mu):.5f} AED")
print(f"Manual Z-Score (Standard Steps): {z_score:.5f}\n")


# Step 3: Manual Calculus Integration Engine (Taylor Series Integration)
# Integration of e^(-Z^2 / 2) from 0 to Z scaled by 1/sqrt(2*pi)
def get_manual_z_area(z):
    # Taylor Series: 1 - Z^2/2 + Z^4/8 - Z^6/48 + Z^8/384 - Z^10/3840
    # Integrating term by term: Z - Z^3/6 + Z^5/40 - Z^7/336 + Z^9/3456 - Z^11/42240
    term1 = z
    term2 = (z ** 3) / 6
    term3 = (z ** 5) / 40
    term4 = (z ** 7) / 336
    term5 = (z ** 9) / 3456
    term6 = (z ** 11) / 42240
    
    # Combined integration result from 0 to Z
    integration_sum = term1 - term2 + term3 - term4 + term5 - term6
    
    # Outside continuous multiplier: 1 / sqrt(2 * pi)
    constant_multiplier = 1 / ((2 * math.pi) ** 0.5)
    
    # Area from center (Z=0) to target Z
    area_from_center = constant_multiplier * integration_sum
    return area_from_center

# Extracting area from center to our specific Z-Score
area_from_center_to_z = get_manual_z_area(z_score)

# Total area to the left of Z (Center left 0.5 + center to Z area)
total_left_area = 0.5 + area_from_center_to_z


# Step 4: Final Probability Area Calculation
# Chance of exceeding target = Total Universe (1.0) - Left Hand Area
probability_exceed = 1.0 - total_left_area
percentage_exceed = probability_exceed * 100

print("============== STEP 4: FINAL RISK PROBABILITY INTEGRATION ==============")
print(f"Area from Center to Z             : {area_from_center_to_z:.5f} (Z-Table equivalent)")
print(f"Probability of being LESS than 250: {total_left_area:.5f} ({total_left_area*100:.2f}%)")
print(f"Probability of EXCEEDING 250 (P>250): {probability_exceed:.5f}")
print(f"Sateek Chance (Final Percentage)  : {percentage_exceed:.2f}%\n")
print("========================================================================")
print(f"VERDICT: Pure {percentage_exceed:.2f}% chance hai ki is 40 bills ke jhund ka average 250 AED se upar nikal jaye.")
print("========================================================================")
```

```python
from scipy import stats

# Given Data
mu, sigma, n, target = 221.55, 279.73, 40, 250.0

# 1. Standard Error & Z-Score Calculation
se = sigma / (n ** 0.5)
z = (target - mu) / se

# 2. Direct Z-Table Reading using Library
# stats.norm.cdf(z) ka matlab hai: Direct Z-Table se 'z' ki value read karna
z_table_value = stats.norm.cdf(z)

# 3. Final Probability (Right Tail Area / Exceeding Chance)
prob_exceed = 1.0 - z_table_value

print(f"Library Z-Score     : {z:.5f}")
print(f"Direct Z-Table Value: {z_table_value:.5f} (Kam aane ka chance)")
print(f"Sateek Chance (P>250): {prob_exceed * 100:.2f}%")
```
