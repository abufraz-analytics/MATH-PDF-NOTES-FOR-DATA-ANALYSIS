```python
# 1. Dice ka total Sample Space (Jo jo aa sakta hai)
dice_sample_space = [1, 2, 3, 4, 5, 6]
total_possible_outcomes = len(dice_sample_space) # Yeh jayega NEECHE (6)

# 2. Variable banate hain favorable outcomes (jo lana hai) ko ginn ne ke liye
favorable_outcomes_count = 0

# 3. Pure Manual Loop: Sample space ke ek-ek element ko check karenge
for outcome in dice_sample_space:
    # Rule: Hamein ya toh number 3 chahiye OR number 5 chahiye
    if outcome == 3 or outcome == 5:
        favorable_outcomes_count += 1 # Agar match hua toh count 1 badhao

# 4. Apply Master Formula: Jo lana hai (Upar) / Jo kul aa sakta hai (Neeche)
manual_probability = favorable_outcomes_count / total_possible_outcomes

print("============== MANUAL ADDITION RESULTS ==============")
print(f"Total Possible Outcomes (Neeche) : {total_possible_outcomes}")
print(f"Favorable Outcomes Found (Upar)   : {favorable_outcomes_count}")
print(f"Calculated Probability           : {manual_probability:.4f}")
print(f"Percentage Chance                : {manual_probability * 100:.2f}%")
```

```python
import random
# 1. Set karte hain ki kitni baar experiment chalana hai
total_trials = 100000
target_hits = 0

# 2. Loop chalayenge 1 lakh baar automatic simulation ke liye
for _ in range(total_trials):
    # random.randint(1, 6) ka matlab hai 1 se 6 tak koi bhi number achanak se aana
    dice_roll = random.randint(1, 6)
    
    # Check karenge: Kya dice par 3 aaya OR 5 aaya?
    if dice_roll == 3 or dice_roll == 5:
        target_hits += 1 # Agar haan, toh hit ginn lo

# 3. Final Probability Calculation
simulation_probability = target_hits / total_trials

print("============== SIMULATION SHORTCUT RESULTS ==============")
print(f"Total Dice Rolls Conducted : {total_trials}")
print(f"Times 3 or 5 Appeared      : {target_hits}")
print(f"Simulation Probability     : {simulation_probability:.4f}")
print(f"Simulation Percentage      : {simulation_probability * 100:.2f}%")
```

```python
# =====================================================================
# DAY 08: MULTIPLICATION RULE PURE MANUAL CODE (Zero Library Way)
# =====================================================================

# 1. Dono Independent Events ke alag-alag Sample Spaces
coin_space = ['Head', 'Tail']
dice_space = [1, 2, 3, 4, 5, 6]

# 2. Variables set karte hain counting ke liye
total_combinations = 0
favorable_combinations = 0

# 3. Nested Loops: Ek-ek karke saare combinations banayenge
for coin_outcome in coin_space:       # Yeh loop 2 baar chalega
    for dice_outcome in dice_space:   # Yeh loop 6 baar chalega
        
        # Har baar ek naya combination banega, toh total count badhao
        total_combinations += 1
        
        # Condition Check: Hamein Sikka par 'Head' AND Paasa par exact 6 chahiye
        if coin_outcome == 'Head' and dice_outcome == 6:
            favorable_combinations += 1

# 4. Master Formula Apply
manual_multiplication_prob = favorable_combinations / total_combinations

# =====================================================================
# RESULTS DISPLAY
# =====================================================================
print("============== MULTIPLICATION MANUAL RESULTS ==============")
print(f"Total Combined Outcomes (Neeche) : {total_combinations}")
print(f"Favorable Combination (Upar)     : {favorable_combinations}")
print(f"Calculated Probability           : {manual_multiplication_prob:.4f}")
print(f"Percentage Chance                : {manual_multiplication_prob * 100:.2f}%")
print("===========================================================")
```

```python
import numpy as np

# =====================================================================
# DAY 08: MULTIPLICATION RULE SHORT CODE (NumPy Matrix Way)
# =====================================================================

# 1. Dono spaces ko numbers mein set karte hain (Sikka: 1=Head, 0=Tail | Paasa: 1 se 6)
coin_options = np.array([1, 0])
dice_options = np.arange(1, 7) # Yeh automatic array bana dega [1, 2, 3, 4, 5, 6]

# 2. NumPy Ka Jaadu: Bina loop ke saare combinations ek sath generate!
coin_matrix, dice_matrix = np.meshgrid(coin_options, dice_options)

# 3. Condition Check: Hamein Sikka par Head (1) AND Paasa par 6 chahiye
# NumPy mein 'AND' ke liye '&' ka use hota hai
target_hits = np.sum((coin_matrix == 1) & (dice_matrix == 6))

# 4. Total outcomes nikalte hain matrix ka size dekh kar
total_combinations = dice_matrix.size

# 5. Master Formula Apply
short_multiplication_prob = target_hits / total_combinations

# =====================================================================
# RESULTS DISPLAY
# =====================================================================
print("============== MULTIPLICATION SHORT RESULTS ==============")
print(f"Total Combined Outcomes (Neeche) : {total_combinations}")
print(f"Favorable Combination (Upar)     : {target_hits}")
print(f"Calculated Percentage            : {short_multiplication_prob * 100:.2f}%")
print("==========================================================")
```
