```python
# Example: 5 Rings Batch mein se exact 2 Rings Kharab hone ka chance

def calculate_factorial(number):
    """
    Step 1: Factorial Engine
    Yeh function kisi bhi number ka factorial (n!) nikalta hai.
    """
    if number == 0 or number == 1:
        return 1
    result = 1
    for i in range(2, number + 1):
        result *= i
    return result


def find_total_conditions(n, k):

    numerator = calculate_factorial(n)
    denominator = calculate_factorial(k) * calculate_factorial(n - k)
    return numerator // denominator


if __name__ == "__main__":
    print("--- PURE MANUAL BINOMIAL ENGINE START ---")
    
    # --- NOTES WALE PARAMETERS INPUT ---
    n = 5       # Total number of rings checked (Batch Size)
    k = 2       # Exact number of defective rings we want to find (Target)
    p = 0.10    # Single ring ke kharab hone ka chance (Defect Rate = 10%)
    q = 1 - p   # Single ring ke sahi hone ka chance (Safe Rate = 90%)

    # --- ENGINE 1: TOTAL CONDITIONS (nCk) ---
    total_paths = find_total_conditions(n, k)
    
    # --- ENGINE 2 & 3: SINGLE CHAIN PROBABILITY (p^k * q^(n-k)) ---
    # target_prob = p^k (Kharab ring ka power)
    target_prob = p ** k
    
    # balance_prob = q^(n-k) (Bachi hui sahi rings ka power)
    balance_prob = q ** (n - k)
    
    # Ek sateek raste ka chance
    single_chain_prob = target_prob * balance_prob

    # --- FINAL ASSEMBLY: CHANNELS MULTIPLIER ---
    # Real Result = Total Conditions * Ek Raaste Ka Chance
    real_result = total_paths * single_chain_prob
    real_percentage = real_result * 100

    # --- SATEEK OUTPUT DISPLAY ---
    print(f"\n[Parameters Matrix]: n={n}, k={k}, p={p}, q={q}")
    print(f"1. Total Conditions Engine (5C2)       : {total_paths} Raaste")
    print(f"2. Single Fixed Chain Chance (p^k * q^n-k): {single_chain_prob:.5f}")
    print(f"3. FINAL REAL RESULT (Probability)    : {real_result:.4f}")
    print(f"4. REAL PERCENTAGE VALUE             : {real_percentage:.2f}%")
```

```python
# =====================================================================
# DAY 13: POISSON DISTRIBUTION CORE ENGINE (PURE MANUAL SOLUTION)
# Example: Showroom mein average 3 customers ke mukable exact 4 aane ka chance
# =====================================================================

def calculate_factorial(number):
    """
    ENGINE 1: Chaos Controller (k!)
    Yeh function target value 'k' ka factorial manually nikalta hai.
    """
    if number == 0 or number == 1:
        return 1
    result = 1
    for i in range(2, number + 1):
        result *= i
    return result


def calculate_exponential_decay(lambda_val):
    """
    ENGINE 2: Kudrat Ka Constant Value Tracker (e^-lambda)
    Euler's number (e = 2.71828) ki power -lambda nikalne ke liye manual setup.
    Formula: 1 / (e^lambda)
    """
    e = 2.71828  # Fixed universal constant value for continuous time/space
    
    # Pehle e ki power lambda nikalte hain (e^lambda)
    e_power_lambda = e ** lambda_val
    
    # Phir use 1 se divide karte hain taaki e^(-lambda) mil sake
    decay_factor = 1 / e_power_lambda
    return decay_factor


if __name__ == "__main__":
    print("--- PURE MANUAL POISSON ENGINE START ---\n")
    
    # --- NOTES WALE PARAMETERS INPUT ---
    lambda_val = 3   # Lambda: Pichle records ka Historical Average Rate (3 VIP customers/hour)
    k = 4            # Target Event: Exact jitne customers ka chance nikalna hai (4 customers)

    # --- BLOCK 1: UMEED KA ENGINE (lambda^k) ---
    # Average rate ki power hamara target. Yeh total volume pressure khada karta hai.
    umeed_engine = lambda_val ** k  # Yani 3^4 = 81

    # --- BLOCK 2: KUDRAT KA BALANCE (e^-lambda) ---
    # System ka base tracker jo probability ko drop hone se balance karta hai.
    kudrat_balance = calculate_exponential_decay(lambda_val)  # Yani 2.71828^-3 ≈ 0.049787

    # --- BLOCK 3: CHAOS CONTROLLER (k!) ---
    # Denominator mein baithkar bikhraav ko normalize karne wala normalizer engine.
    chaos_controller = calculate_factorial(k)  # Yani 4! = 24

    # --- FINAL ASSEMBLY: THE POISSON EQUATION ---
    # Formula: (lambda^k * e^-lambda) / k!
    # Umeed ka engine aur Kudrat ka balance multiply honge, phir Chaos controller se divide honge.
    real_result = (umeed_engine * kudrat_balance) / chaos_controller
    real_percentage = real_result * 100

    # --- SATEEK OUTPUT DISPLAY ENGINE ---
    print(f"[Parameters Loaded]: Lambda (Average)={lambda_val}, k (Target)={k}")
    print(f"1. Umeed Ka Engine (3^4)              : {umeed_engine}")
    print(f"2. Kudrat Ka Balance (e^-3)           : {kudrat_balance:.6f}")
    print(f"3. Chaos Controller Normalizer (4!)   : {chaos_controller}")
    print(f"4. 🏆 FINAL REAL RESULT (Probability) : {real_result:.4f}")
    print(f"5. 📈 REAL PERCENTAGE VALUE          : {real_percentage:.2f}%")
    print("\n--- ENGINE EXECUTION SUCCESSFUL ---")
```
