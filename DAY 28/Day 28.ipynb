```python
import sympy as sp

# 1. Python ko bataya ki 'm' ek math ka variable (x, y jaisa) hai
m = sp.Symbol('m')

# 2. Humne code ko bina derivative nikale seedhe asli equations de di
log = 2 * m + 3
items = 3 * (log ** 2) + 2
profit = 5 * (items ** 2) + 100

# 3. YAHAN HO RAHA HAI ASLI CALCULUS! 
# sp.diff ka matlab hai: "Python, tum khud iska Derivative (Calculus) karo"
asli_calculus_derivative = sp.diff(profit, m)

# 4. Ab m = 2 rakh kar answer nikal lo
final_jump = asli_calculus_derivative.subs(m, 2)

print(f"Calculus ke baad formula bana: {asli_calculus_derivative}")
print(f"Final Answer m=2 par: {final_jump}")
```
