```python
# 1. Hamaare paas 9 customers ka gold purchase data hai
gold_purchases = [35, 12, 20, 15, 10, 22, 30, 18, 25]

# Step 1: Data ko ascending order me sort karna
gold_purchases.sort()
print("Sorted Data:", gold_purchases)
# Output: [10, 12, 15, 18, 20, 22, 25, 30, 35]

n = len(gold_purchases)  # n = 9 (Odd)

# Step 2: Median (Q2) ki index aur value nikalna
# Odd data me single center index hoti hai: (9 // 2) = 4th index
median_index = n // 2
q2 = gold_purchases[median_index]
print(f"Median (Q2): {q2} (Index: {median_index})")

# Step 3: Median ko hata kar Lower aur Upper Halves banana
# Median (index 4) ke pehle ka data [0 to 3]
lower_half = gold_purchases[:median_index]
# Median (index 4) ke baad ka data [5 to end]
upper_half = gold_purchases[median_index + 1 :]

print("Lower Half:", lower_half)  # [10, 12, 15, 18]
print("Upper Half:", upper_half)  # [22, 25, 30, 35]

# Step 4: Q1 Nikalna (Lower Half ka Median)
# lower_half me 4 elements hain (Even length). 
# Iska median hoga beech ke do elements ka average: (12 + 15) / 2
m_low = len(lower_half)
idx1 = m_low // 2 - 1  # 1st index (12)
idx2 = m_low // 2      # 2nd index (15)
q1 = (lower_half[idx1] + lower_half[idx2]) / 2

# Step 5: Q3 Nikalna (Upper Half ka Median)
# upper_half me bhi 4 elements hain. 
# Iska median hoga beech ke do elements ka average: (25 + 30) / 2
m_up = len(upper_half)
idx3 = m_up // 2 - 1   # 1st index (25)
idx4 = m_up // 2       # 2nd index (30)
q3 = (upper_half[idx3] + upper_half[idx4]) / 2

# Step 6: IQR Formula lagana
iqr = q3 - q1

print("--- FINAL LOGIC RESULTS ---")
print("Q1 (25th Percentile):", q1)
print("Q3 (75th Percentile):", q3)
print("Interquartile Range (IQR):", iqr)
```

```python
# Hamaare paas 10 customers ka gold purchase data hai
gold_purchases_even = [35, 12, 20, 15, 10, 22, 30, 18, 25, 40]

# Step 1: Data ko ascending order me sort karna
gold_purchases_even.sort()
print("Sorted Even Data:", gold_purchases_even)
# Output: [10, 12, 15, 18, 20, 22, 25, 30, 35, 40]

n = len(gold_purchases_even)  # n = 10 (Even)

# Step 2: Pure data ka Median (Q2) nikalna
# 10 elements me index 4 (20) aur index 5 (22) center hote hain
mid1 = n // 2 - 1  # Index 4
mid2 = n // 2      # Index 5
q2 = (gold_purchases_even[mid1] + gold_purchases_even[mid2]) / 2
print(f"Overall Median (Q2): {q2}")  # (20 + 22) / 2 = 21.0

# Step 3: Data ko exact do barabar hisson me baantna (Pehle 5 aur Aakhiri 5)
# Index 0 se 4 tak (5 elements)
lower_half = gold_purchases_even[: n // 2]
# Index 5 se end tak (5 elements)
upper_half = gold_purchases_even[n // 2 :]

print("Lower Half (First 5):", lower_half)  # [10, 12, 15, 18, 20]
print("Upper Half (Last 5):", upper_half)    # [22, 25, 30, 35, 40]

# Step 4: Q1 Nikalna (Lower Half ka Median)
# lower_half me 5 elements hain (Odd count).
# 5 elements ka perfect center hota hai index 2.
m_low = len(lower_half)              # 5
q1_index = m_low // 2                # 5 // 2 = 2 (Index 2)
q1 = lower_half[q1_index]            # Index 2 par baitha hai 15

# Step 5: Q3 Nikalna (Upper Half ka Median)
# upper_half me bhi 5 elements hain (Odd count).
# Iska bhi perfect center hoga index 2.
m_up = len(upper_half)               # 5
q3_index = m_up // 2                 # 5 // 2 = 2 (Index 2)
q3 = upper_half[q3_index]            # Index 2 par baitha hai 30

# Step 6: IQR Formula
iqr = q3 - q1

print("--- FINAL EVEN LOGIC RESULTS ---")
print("Q1 (25th Percentile):", q1)   # Output: 15.0
print("Q3 (75th Percentile):", q3)   # Output: 30.0
print("Interquartile Range (IQR):", iqr)  # Output: 15.0
```

```python
# Maaniye hamare paas 7 customers ka gold purchase data hai
gold_sales = [10, 20, 30, 20, 30, 10, 20]

# Step 1: Data ko sort karna (Median aur Mode ke liye zaroori hai)
gold_sales.sort()
print("Sorted Data:", gold_sales)
# Output: [10, 10, 20, 20, 20, 30, 30]


# --- 1. MANUAL MEAN (AVERAGE) ---
# Saare numbers ka sum divided by total elements
total_sum = sum(gold_sales)
n = len(gold_sales)
manual_mean = total_sum / n


# --- 2. MANUAL MEDIAN (MIDDLE VALUE) ---
# Kyunki 7 elements hain (Odd), toh perfect center hoga index (7 // 2) = 3
median_index = n // 2
manual_median = gold_sales[median_index]


# --- 3. MANUAL MODE (MOST FREQUENT VALUE) ---
# Hum har number ki frequency (ginti) nikalenge aur check karenge kaun sabse zyada aaya
frequency_dict = {}
for item in gold_sales:
    if item in frequency_dict:
        frequency_dict[item] += 1
    else:
        frequency_dict[item] = 1

# Sabse badi ginti (max count) aur uske number ko dhoondhna
max_count = 0
manual_mode = None
for number, count in frequency_dict.items():
    if count > max_count:
        max_count = count
        manual_mode = number


# --- 4. MANUAL SYMMETRY CHECK LOGIC ---
print("\n--- MANUAL STATS CALCULATIONS ---")
print(f"Mean  : {manual_mean}")
print(f"Median: {manual_median}")
print(f"Mode  : {manual_mode}")

print("\n--- SHAPE OF DATA RESULT ---")
if manual_mean == manual_median == manual_mode:
    print("Result: Data perfectly SYMMETRIC hai! (Bell Curve banega)")
else:
    print("Result: Data symmetric nahi hai, ye skewed hai.")
```

```python
import statistics
from scipy import stats

# Hamaara wahi data
gold_sales = [10, 20, 30, 20, 30, 10, 20]

# 1. Direct ek-ek line mein Mean, Median, Mode nikalna
mean_val = statistics.mean(gold_sales)
median_val = statistics.median(gold_sales)
mode_val = statistics.mode(gold_sales)  # Ya fir stats.mode(gold_sales)[0]

print("--- DIRECT LIBRARIES RESULTS ---")
print(f"Mean  : {mean_val}")    # Output: 20.0
print(f"Median: {median_val}")  # Output: 20
print(f"Mode  : {mode_val}")    # Output: 20

# 2. Symmetry Check Rule
print("\n--- SHAPE OF DATA RESULT ---")
if mean_val == median_val == mode_val:
    print("Result: Data perfectly SYMMETRIC hai!")
else:
    print("Result: Data skewed hai.")
```

```python
# 1. Showroom data: Ek customer ne achanak 80 grams gold liya (Outlier)
gold_sales_positive = [10, 15, 12, 15, 10, 15, 80]

# Step 1: Data ko sort karna
gold_sales_positive.sort()
print("Sorted Positive Skewed Data:", gold_sales_positive)
# Output: [10, 10, 12, 15, 15, 15, 80]

n = len(gold_sales_positive)  # n = 7

# --- 1. MANUAL MEAN ---
manual_mean = sum(gold_sales_positive) / n
# Sum = 157, Mean = 157 / 7 = 22.42


# --- 2. MANUAL MEDIAN ---
# 7 items me perfect center index = 3
median_index = n // 2
manual_median = gold_sales_positive[median_index]
# Index 3 par baitha hai 15


# --- 3. MANUAL MODE (Wahi ginti wala logic) ---
frequency_dict = {}
for item in gold_sales_positive:
    if item in frequency_dict:
        frequency_dict[item] += 1
    else:
        frequency_dict[item] = 1

max_count = 0
manual_mode = None
for number, count in frequency_dict.items():
    if count > max_count:
        max_count = count
        manual_mode = number
# 15 teen baar aaya hai, toh Mode = 15


# --- 4. SKEWNESS CHECK LOGIC ---
print("\n--- MANUAL CALCULATIONS ---")
print(f"Mean  : {round(manual_mean, 2)}")  # 22.42
print(f"Median: {manual_median}")          # 15
print(f"Mode  : {manual_mode}")            # 15

print("\n--- SHAPE OF DATA RESULT ---")
if manual_mean > manual_median:
    print("Result: Data POSITIVELY SKEWED (Right-Skewed) hai!")
    print("Logic: Bade customer (80) ne Mean ko khinch kar bada kar diya!")
elif manual_mean < manual_median:
    print("Result: Data NEGATIVELY SKEWED (Left-Skewed) hai!")
else:
    print("Result: Data Symmetric hai!")
```
