```python
import numpy as np

# =================================================================================
# 🏢 SCENARIO / QUESTION:
# Goda Gold Dubai showroom handles four customer segments daily:
# x1 = Retail Customers, x2 = Corporate Buyers, x3 = VIP Investors, x4 = Tourist Buyers
#
# Four dynamic metrics are tracked across these segments:
# 1. Total Gold Items Bought:          2*x1 + 10*x2 + 25*x3 + 5*x4   = 160 items
# 2. Total Making Charge Revenue(AED): 100*x1 + 500*x2 + 2000*x3 + 300*x4 = 9900 AED
# 3. Total Loyalty Points Issued:      1*x1 + 4*x2 + 10*x3 + 2*x4     = 66 points
# 4. Total Showroom Time Spent (Min):  15*x1 + 60*x2 + 120*x3 + 30*x4 = 930 minutes
#
# TASK: Write an automated Python engine using libraries to instantly extract 
# the Determinant, Inverse Matrix, and the Final Segment Values (x1, x2, x3, x4).
# =================================================================================

print("======= GODA GOLD AUTOMATED ACCOUNTING ENGINE =======")

# Step 1: Coefficients Matrix (A) aur Targets Vector (B) define karein
A = np.array([
    [2, 10, 25, 5],
    [100, 500, 2000, 300],
    [1, 4, 10, 2],
    [15, 60, 120, 30]
])

B = np.array([160, 9900, 66, 930])

# Step 2: Library functions ka use karke Determinant aur Inverse nikalein
det_A = np.linalg.det(A)
A_inverse = np.linalg.inv(A)

# Step 3: Pure system ko solve karein (X = A_inverse * B)
X_solutions = np.linalg.solve(A, B)

# ======= PRINTING DETAILED OUTPUTS INSTANTLY =======
print(f"\n1. DETERMINANT (|A|): {round(det_A, 2)}")

print("\n2. INVERSE MATRIX (A^-1):")
# Clean formatting ke liye decimal places ko 4 digits par round kiya hai
print(np.round(A_inverse, 4))

print("\n3. FINAL SEGMENT ANALYSIS SOLUTIONS (X):")
customer_types = [
    "Retail Customers (x1)", 
    "Corporate Buyers (x2)", 
    "VIP Investors (x3)", 
    "Tourist Buyers (x4)"
]

for i, customer in enumerate(customer_types):
    print(f" -> {customer} = {round(X_solutions[i])}")

print("=====================================================")
```
