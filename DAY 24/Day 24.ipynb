```python
# ====================================================
# DAY 24: GRADIENT DESCENT COMPLETE LOOP (CASE 1)
# ====================================================

# 1. Hamara sateek data point [cite: 38]
X = 2
actual_y = 4

# 2. Initial Random Guess [cite: 40]
m = 1.0
c = 0.5

# 3. Learning Rate [cite: 49]
learning_rate = 0.1

# Iterations ginne ke liye counter
loop_count = 0

print("Loop Shuru Ho Raha Hai...\n")
print(f"Shuruati Guess -> m = {m}, c = {c}\n")
print("-" * 50)

# Yeh loop tab tak chalega jab tak dhalan 0 ke bohot paas nahi pahunch jaati 
while True:
    loop_count += 1
    
    # STEP 1: Prediction nikalna [cite: 41]
    pred_y = (m * X) + c
    
    # STEP 2: Error Score (MSE) [cite: 42, 43]
    error = (actual_y - pred_y) ** 2
    
    # STEP 3: Calculus Dhalan (Gradients) [cite: 44, 45, 47]
    gradient_m = -2 * X * (actual_y - pred_y)
    gradient_c = -2 * (actual_y - pred_y)
    
    # Iteration ki report screen par print karna
    print(f"Iteration {loop_count}:")
    print(f"   Predicted Y = {pred_y:.4f} | Total Error = {error:.4f}")
    print(f"   Dhalan_m = {gradient_m:.4f} | Dhalan_c = {gradient_c:.4f}")
    
    # STOPPING CRITERIA: Agar dono dhalan 0 ke ekdum paas pahunch gayi, toh jackpot! 
    # abs() ka matlab hai minus sign ignore karke sirf value check karna
    if abs(gradient_m) < 0.0001 and abs(gradient_c) < 0.0001:
        print("-" * 50)
        print("JACKPOT! Dhalan ekdum Flat (0) ho gayi hai. Loop ko roka ja raha hai! ")
        break
        
    # STEP 4: Simultaneous Update (Dono ko ek sath badalna) 
    m = m - (learning_rate * gradient_m)
    c = c - (learning_rate * gradient_c)
    
    print(f"   Naye Guess -> m = {m:.4f}, c = {c:.4f}\n")

print("-" * 50)
print("FINAL RESULTS:")
print(f"Total Loops Chale: {loop_count}")
print(f"Computer dwara dhoodha gaya sateek m = {m:.2f}")
print(f"Computer dwara dhoodha gaya sateek c = {c:.2f}")
print(f"Final Equation Bani: Y = {m:.2f}*X + {c:.2f}")
```

```python
# ====================================================
# DAY 24: MULTI-DAY LOOP - EXPLODING GRADIENT FIXED
# ====================================================

# 1. Multi-Day Dataset
X_data = [2, 4]
actual_y_data = [4, 12]

# 2. Initial Balanced Guess
m = 2.2
c = 0.5

# 3. Learning Rate KO CHOTA KIYA (0.1 se 0.01 kiya taaki overshoot na ho)
learning_rate = 0.01
loop_count = 0

print("Multi-Day Data Loop Shuru Ho Raha Hai...\n")
print(f"Shuruati Guess -> m = {m}, c = {c}\n")
print("-" * 60)

while True:
    loop_count += 1
    
    total_error = 0
    sum_gradient_m = 0
    sum_gradient_c = 0
    
    for i in range(len(X_data)):
        X = X_data[i]
        actual_y = actual_y_data[i]
        
        # STEP 1: Prediction
        pred_y = (m * X) + c
        
        # STEP 2: Error Score
        error = (actual_y - pred_y) ** 2
        total_error += error
        
        # STEP 3: Gradients
        grad_m = -2 * X * (actual_y - pred_y)
        grad_c = -2 * (actual_y - pred_y)
        
        sum_gradient_m += grad_m
        sum_gradient_c += grad_c

    # Average nikaalna
    mse_cost = total_error / len(X_data)
    avg_gradient_m = sum_gradient_m / len(X_data)
    avg_gradient_c = sum_gradient_c / len(X_data)
    
    # STOPPING CRITERIA: Jab dono Average Gradients zero ke paas pahunchein
    if abs(avg_gradient_m) < 0.0001 and abs(avg_gradient_c) < 0.0001:
        print("-" * 60)
        print("JACKPOT! Dono dinon ka data globally balance ho gaya. Loop Stop!")
        break
        
    # STEP 4: Simultaneous Update
    m = m - (learning_rate * avg_gradient_m)
    c = c - (learning_rate * avg_gradient_c)

print("-" * 60)
print("FINAL GLOBAL RESULTS:")
print(f"Total Loops Chale: {loop_count}")
print(f"Computer ka sateek balanced m = {m:.2f}")
print(f"Computer ka sateek balanced c = {c:.2f}")
print(f"Final Global Equation: Y = {m:.2f}*X + ({c:.2f})")
```
