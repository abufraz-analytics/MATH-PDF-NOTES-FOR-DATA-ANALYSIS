```python
print("--- STEP 1: Observed Data (Jo Table Mein Diya Hai) ---")
# Humare paas 4 cells hain table mein:
# Row 1 (Tourist): Rings = 50, Chains = 150
# Row 2 (Local): Rings = 150, Chains = 50

t_rings = 50
t_chains = 150
l_rings = 150
l_chains = 50

print("Tourist Rings:", t_rings, "| Tourist Chains:", t_chains)
print("Local Rings:", l_rings, "| Local Chains:", l_chains)


print("\n--- STEP 2: Totals Nikalna (Plus Karke) ---")
# Row ke totals
tourist_total = t_rings + t_chains      # 50 + 150 = 200
local_total = l_rings + l_chains        # 150 + 50 = 200

# Column ke totals
rings_total = t_rings + l_rings        # 50 + 150 = 200
chains_total = t_chains + l_chains      # 150 + 50 = 200

# Sabka total (Grand Total)
grand_total = tourist_total + local_total  # 200 + 200 = 400

print("Tourist Total Row:", tourist_total)
print("Local Total Row:", local_total)
print("Rings Total Column:", rings_total)
print("Chains Total Column:", chains_total)
print("Grand Total (N):", grand_total)


print("\n--- STEP 3: Degree of Freedom (df) ---")
# Formula: (Rows - 1) * (Columns - 1)
# Humaray paas 2 Rows aur 2 Columns hain
df = (2 - 1) * (2 - 1)                  # 1 * 1 = 1
print("Degree of Freedom (df) =", df)


print("\n--- STEP 4: Expected Values Nikalna ---")
# Formula: (Row Total * Column Total) / Grand Total
# Chunki saari values 200 hain, toh sabka expected same aayega

exp_t_rings = (tourist_total * rings_total) / grand_total    # (200 * 200) / 400 = 100
exp_t_chains = (tourist_total * chains_total) / grand_total  # (200 * 200) / 400 = 100
exp_l_rings = (local_total * rings_total) / grand_total      # (200 * 200) / 400 = 100
exp_l_chains = (local_total * chains_total) / grand_total    # (200 * 200) / 400 = 100

print("Expected Tourist Rings:", exp_t_rings)
print("Expected Tourist Chains:", exp_t_chains)
print("Expected Local Rings:", exp_l_rings)
print("Expected Local Chains:", exp_l_chains)


print("\n--- STEP 5: Chi-Square Score (χ²) ---")
# Formula: (Observed - Expected)² / Expected
# Har ek cell ke liye alag-alag nikalenge aur baad mein plus karenge

part1 = ((t_rings - exp_t_rings) ** 2) / exp_t_rings    # (50 - 100)² / 100 = 2500 / 100 = 25
part2 = ((t_chains - exp_t_chains) ** 2) / exp_t_chains  # (150 - 100)² / 100 = 2500 / 100 = 25
part3 = ((l_rings - exp_l_rings) ** 2) / exp_l_rings    # (150 - 100)² / 100 = 2500 / 100 = 25
part4 = ((l_chains - exp_l_chains) ** 2) / exp_l_chains  # (50 - 100)² / 100 = 2500 / 100 = 25

# Sabko plus kar do
chi_square_score = part1 + part2 + part3 + part4        # 25 + 25 + 25 + 25 = 100
print("Final Calculated Chi-Square Score =", chi_square_score)


print("\n--- STEP 6: Percentage Nikalna (Library Se) ---")
from scipy import stats

# 1. Humare Table Data ke liye (df=1, Chi-Square=100)
p_value_table = stats.chi2.sf(chi_square_score, df)
percentage_table = p_value_table * 100
print("Table Data Percentage:", percentage_table, "%")
```
