```python
# =========================================================================
# MACHINE DEFINITION: Pehle hum probability nikaalne ki machine bana rahe hain
# =========================================================================

def calculate_basic_probability(favorable_outcomes, total_outcomes):
    """
    def ka matlab hai hum ek function (machine) bana rahe hain.
    Is machine ko do numbers chahiye: favorable_outcomes aur total_outcomes.
    """
    # Safety Check 1: Agar total outcomes 0 ya minus mein ho, toh error de do
    if total_outcomes <= 0:
        return "Total outcomes must be greater than 0."
    
    # Safety Check 2: Jo lana hai (favorable) woh total se bada nahi ho sakta
    if favorable_outcomes > total_outcomes:
        return "Favorable outcomes cannot exceed total outcomes."
    
    # STEP 1: Basic Master Formula -> P(E) = Favorable / Total
    probability_event = favorable_outcomes / total_outcomes
    
    # STEP 2: Complement Rule (Na hone ka chance) -> P(E') = 1 - P(E)
    probability_complement = 1 - probability_event
    
    # STEP 3: Percentage mein badalna (100 se guna karke)
    prob_percentage = probability_event * 100
    comp_percentage = probability_complement * 100
    
    # STEP 4: Saare answers ko round off karke (point ke baad 4 aur 2 digit tak) return karna
    return {
        "Probability of Event P(E)": round(probability_event, 4),
        "Probability Percentage": f"{round(prob_percentage, 2)}%",
        "Complement Probability P(E')": round(probability_complement, 4),
        "Complement Percentage": f"{round(comp_percentage, 2)}%"
    }


# =========================================================================
# ACTUAL EXECUTION: Ab hum questions ka data machine mein daal kar run karenge
# =========================================================================

print("=" * 60)
# 📌 QUESTION 1:
# Ek pure jewelry box mein 12 Rings hain, jisme se 4 Diamond Rings hain 
# aur 8 Gold Rings hain. Agar aap bina dekhe ek ring uthate hain, 
# toh Diamond ring aane ka chance kya hai?
print("RUNNING QUESTION 1: Diamond Ring Inside Jewelry Box")
print("=" * 60)

# Solution Logic: Favorable (Diamond) = 4, Total Rings = 12
q1_result = calculate_basic_probability(favorable_outcomes=4, total_outcomes=12)

# Screen par output dikhane ke liye loop
for key, value in q1_result.items():
    print(f"{key} : {value}")


print("\n" + "=" * 60)
# 📌 QUESTION 2:
# Ek financial ledger register mein total 50 pages hain. Unme se 15 pages 
# par calculation error hai. Agar manager randomly koi bhi ek page kholta hai, 
# toh us page par error NA hone (Complement Event) ka chance kya hai?
print("RUNNING QUESTION 2: Financial Ledger Error Audit")
print("=" * 60)

# Solution Logic: Total Pages = 50, Error Pages = 15
# Error NA hone waale pages (Favorable) = 50 - 15 = 35 pages
q2_result = calculate_basic_probability(favorable_outcomes=35, total_outcomes=50)

# Screen par output dikhane ke liye loop
for key, value in q2_result.items():
    print(f"{key} : {value}")

print("=" * 60)
```

```python
# =========================================================================
# MACHINE DEFINITION: Addition Rule (Mutually Exclusive) ki manual machine
# =========================================================================

def calculate_addition_exclusive(prob_A, prob_B):
    """
    Bina kisi external library ke, mutually exclusive events ke liye 
    Addition Rule [P(A or B) = P(A) + P(B)] ka manual implementation.
    """
    # Safety Check: Probability kabhi minus mein ya 1 se badi nahi ho sakti
    if not (0 <= prob_A <= 1) or not (0 <= prob_B <= 1):
        return "Error: Individual probabilities must be between 0 and 1."
        
    # Safety Check: Mutually exclusive mein dono ka jod bhi 1 se bada nahi hona chahiye
    if (prob_A + prob_B) > 1:
        return "Error: Sum of mutually exclusive probabilities cannot exceed 100%."

    # ASLI LOGIC: OR / Ya ke khel mein dono ko plus (+) karte hain
    total_probability = prob_A + prob_B
    
    # Percentage calculation
    total_percentage = total_probability * 100
    
    return {
        "Probability of Event A P(A)": round(prob_A, 4),
        "Probability of Event B P(B)": round(prob_B, 4),
        "Total Probability P(A or B)": round(total_probability, 4),
        "Total Percentage": f"{round(total_percentage, 2)}%"
    }


# =========================================================================
# ACTUAL EXECUTION: Ab hum questions ka data machine mein daal kar run karenge
# =========================================================================

print("=" * 60)
# 📌 QUESTION 3:
# Ek standard Ludo dice (paasa) feka jata hai. Probability nikaliye ki 
# upar aane waala number ya toh ek Even Number (2, 4, 6) ho YA fir number 5 ho.
print("RUNNING QUESTION 3: Standard Ludo Dice (Even OR 5)")
print("=" * 60)

# Solution Logic: 
# Total outcomes = 6
# Even numbers (2,4,6) = 3 outcomes -> P(Even) = 3/6 = 0.5
# Number 5 = 1 outcome -> P(5) = 1/6 = 0.1667
p_even = 3 / 6
p_five = 1 / 6

q3_result = calculate_addition_exclusive(prob_A=p_even, prob_B=p_five)

# Hamara pyara loop output dikhane ke liye
for key, value in q3_result.items():
    print(f"{key} : {value}")


print("\n" + "=" * 60)
# 📌 QUESTION 4:
# Ek gold showroom mein ek customer aata hai. Uske 'Gold Chain' khareodne ka chance 
# 25% (0.25) hai, aur 'Bangles' khareodne ka chance 15% (0.15) hai. Maan lijiye 
# woh dono mein se koi ek hi cheez khareod sakta hai. Kya chance hai ki woh ya toh Chain YA Bangles khareodega?
print("RUNNING QUESTION 4: Showroom Customer (Chain OR Bangles)")
print("=" * 60)

# Solution Logic: P(Chain) = 0.25, P(Bangles) = 0.15
p_chain = 0.25
p_bangles = 0.15

q4_result = calculate_addition_exclusive(prob_A=p_chain, prob_B=p_bangles)

for key, value in q4_result.items():
    print(f"{key} : {value}")

print("=" * 60)
```

```python
# =========================================================================
# MACHINE DEFINITION: Non-Mutually Exclusive (Overlap wale) events ki machine
# =========================================================================

def calculate_addition_non_exclusive(prob_A, prob_B, prob_both):
    """
    Dono events mein jo common hissa (overlap) hai, use ek baar minus 
    karne ka manual logic: P(A or B) = P(A) + P(B) - P(A and B)
    """
    # Safety Check: Saari individual probabilities 0 se 1 ke beech honi chahiye
    if not (0 <= prob_A <= 1) or not (0 <= prob_B <= 1) or not (0 <= prob_both <= 1):
        return "Error: Probabilities must be between 0 and 1."
        
    # Safety Check: Common hissa kabhi bhi kisi single event se bada nahi ho sakta
    if prob_both > prob_A or prob_both > prob_B:
        return "Error: Common probability cannot be greater than individual probabilities."

    # ASLI LOGIC: Dono ko jodo (+) aur common wale ko EK BAAR ghatao (-)
    total_probability = prob_A + prob_B - prob_both
    
    # Percentage calculation
    total_percentage = total_probability * 100
    
    return {
        "Probability of Event A P(A)": round(prob_A, 4),
        "Probability of Event B P(B)": round(prob_B, 4),
        "Common Overlap P(A and B)": round(prob_both, 4),
        "Total Probability P(A or B)": round(total_probability, 4),
        "Total Percentage": f"{round(total_percentage, 2)}%"
    }


# =========================================================================
# ACTUAL EXECUTION: Questions ka data machine mein daal kar run karenge
# =========================================================================

print("=" * 65)
# 📌 QUESTION 5:
# Ek 52 taash ke patton (cards) ki gaddi se ek patta nikala jata hai. 
# Probability nikaliye ki woh patta ya toh ek 'Kala Patta' (Black Card) ho 
# YA fir ek 'Badshah' (King) ho.
print("RUNNING QUESTION 5: Playing Cards (Black Card OR King)")
print("=" * 65)

# Solution Logic:
# Total Cards = 52
# Total Black Cards = 26  -> P(Black) = 26/52 = 0.5
# Total Kings = 4         -> P(King) = 4/52 = 0.0769
# Common (Kala Badshah) = 2 patte aise hain jo Black bhi hain aur King bhi hain!
# P(Black and King) = 2/52 = 0.0385

p_black = 26 / 52
p_king = 4 / 52
p_black_king = 2 / 52

q5_result = calculate_addition_non_exclusive(prob_A=p_black, prob_B=p_king, prob_both=p_black_king)

# Hamara vahi pyaara solid loop output dikhane ke liye
for k, v in q5_result.items():
    print(f"{k} : {v}")


print("\n" + "=" * 65)
# 📌 QUESTION 6:
# Ek jewelry outlet mein audit ho raha hai. Total 100 items hain. 
# 30 items par 'Tagging Error' hai, 20 items par 'Weight Mismatch' hai, 
# aur 10 items aise hain jinme Error aur Mismatch DONO hain. 
# Agar manager ek item uthaye, toh usme koi bhi EK kharabi (Error YA Mismatch) hone ka chance kya hai?
print("RUNNING QUESTION 6: Jewelry Inventory Audit Overlap")
print("=" * 65)

# Solution Logic: Total = 100
# P(Tagging Error) = 30/100 = 0.3
# P(Weight Mismatch) = 20/100 = 0.2
# P(Both) = 10/100 = 0.1

p_tag = 30 / 100
p_weight = 20 / 100
p_both_errors = 10 / 100

q6_result = calculate_addition_non_exclusive(prob_A=p_tag, prob_B=p_weight, prob_both=p_both_errors)

for k, v in q6_result.items():
    print(f"{k} : {v}")

print("=" * 65)
```

```python
# =========================================================================
# MACHINE DEFINITION: Multiplication Rule (Independent) ki manual machine
# =========================================================================

def calculate_multiplication_independent(prob_A, prob_B):
    """
    Bina kisi library ke, do aazaad (independent) events ke ek sath 
    ya ek ke baad ek hone ki probability: P(A and B) = P(A) * P(B)
    """
    # Safety Check: Probability hamesha 0 se 1 ke beech honi chahiye
    if not (0 <= prob_A <= 1) or not (0 <= prob_B <= 1):
        return "Error: Probabilities must be between 0 and 1."

    # ASLI LOGIC: "AND" / "Aur" ke khel mein hamesha Guna (Multiply *) hota hai
    total_probability = prob_A * prob_B
    
    # Percentage mein badalna
    total_percentage = total_probability * 100
    
    return {
        "Probability of Event A P(A)": round(prob_A, 4),
        "Probability of Event B P(B)": round(prob_B, 4),
        "Combined Probability P(A and B)": round(total_probability, 4),
        "Total Percentage": f"{round(total_percentage, 2)}%"
    }


# =========================================================================
# ACTUAL EXECUTION: Data ko machine mein daal kar run karenge
# =========================================================================

print("=" * 65)
# 📌 QUESTION 7:
# Aap ek sath do kaam karte hain: Ek sikka (Coin) uchhalte hain aur ek Ludo Dice fekte hain.
# Kya probability hai ki sikke par 'Head' aaye AUR dice par number '6' aaye?
print("RUNNING QUESTION 7: Coin (Head) AND Dice (6)")
print("=" * 65)

# Solution Logic:
# Coin par Head aane ka chance: 1/2 = 0.5
# Dice par 6 aane ka chance: 1/6 = 0.1667
p_head = 1 / 2
p_six = 1 / 6

q7_result = calculate_multiplication_independent(prob_A=p_head, prob_B=p_six)

# Hamara standard loop
for k, v in q7_result.items():
    print(f"{k} : {v}")


print("\n" + "=" * 65)
# 📌 QUESTION 8:
# Ek jewelry manufacturing unit mein do alag-alag machines hain. 
# Machine 1 se koi piece kharab (defect) nikalne ka chance 5% (0.05) hai. 
# Machine 2 se kharab nikalne ka chance 10% (0.10) hai. 
# Agar dono se ek-ek piece check kiya jaye, toh kya chance hai ki DONO hi pieces kharab (Defect) hon?
print("RUNNING QUESTION 8: Independent Factory Machine Defects")
print("=" * 65)

# Solution Logic: P(Machine 1 Defect) = 0.05, P(Machine 2 Defect) = 0.10
p_m1_defect = 0.05
p_m2_defect = 0.10

q8_result = calculate_multiplication_independent(prob_A=p_m1_defect, prob_B=p_m2_defect)

for k, v in q8_result.items():
    print(f"{k} : {v}")

print("=" * 65)
```

```python
# =========================================================================
# MACHINE DEFINITION: Dependent Events (Bina waapas rakhe) ki manual machine
# =========================================================================

def calculate_multiplication_dependent(favorable, total):
    """
    Bina kisi library ke, dependent events ka manual logic.
    Pehla piece nikaalne ke baad total aur favorable dono mein se 1-1 ghatt jata hai.
    """
    # 1. PEHLA CHANCE: P(A)
    prob_A = favorable / total
    
    # 2. DOOSRA CHANCE: P(B|A) -> Ek piece nikal gaya, toh dono mein se 1-1 kam ho gaya
    fav_after_first = favorable - 1
    total_after_first = total - 1
    
    prob_B_given_A = fav_after_first / total_after_first
    
    # ASLI LOGIC: Dono chances ko aapas mein Guna (*) karna
    total_probability = prob_A * prob_B_given_A
    total_percentage = total_probability * 100
    
    return {
        "First Draw Probability P(A)": round(prob_A, 4),
        "Second Draw Probability P(B|A)": round(prob_B_given_A, 4),
        "Combined Probability (Both)": round(total_probability, 4),
        "Total Percentage": f"{round(total_percentage, 2)}%"
    }


# =========================================================================
# ACTUAL EXECUTION: Data ko machine mein daal kar run karenge
# =========================================================================

print("=" * 70)
# 📌 QUESTION 9:
# Ek thaili mein 10 leather ki boles (balls) hain, jisme se 4 'Lal' (Red) hain. 
# Aap bina dekhe ek ball nikaalte hain (aur use waapas nahi rakhte). Phir aap doosri ball nikaalte hain.
# Kya probability hai ki DONO hi balls 'Lal' (Red) niklein?
print("RUNNING QUESTION 9: Bag of Balls (Red AND Red - Without Replacement)")
print("=" * 70)

# Solution Logic: Total = 10, Favorable (Red) = 4
# Pehli baar: 4/10
# Doosri baar: Ek red kam hui (3) aur ek total kam hua (9) -> 3/9
q9_result = calculate_multiplication_dependent(favorable=4, total=10)

# Hamara standard loop
for k, v in q9_result.items():
    print(f"{k} : {v}")


print("\n" + "=" * 70)
# 📌 QUESTION 10:
# Ek jewelry box mein 6 Gold Rings hain aur total 15 rings hain. 
# Ek ke baad ek, do rings bina waapas rakhe uthayi jaati hain audit ke liye.
# Kya chance hai ki DONO ki dono rings 'Gold' ki hi niklein?
print("RUNNING QUESTION 10: Jewelry Box Audit (Gold AND Gold)")
print("=" * 70)

# Solution Logic: Total = 15, Favorable (Gold) = 6
# Pehli baar: 6/15
# Doosri baar: Ek gold kam hui (5) aur ek total kam hua (14) -> 5/14
q10_result = calculate_multiplication_dependent(favorable=6, total=15)

for k, v in q10_result.items():
    print(f"{k} : {v}")

print("=" * 70)
```

```python
from fractions import Fraction

# Short Machine: Initial favorable aur total lo, aur bina waapas rakhe guna (*) kar do
def short_multiplication_dependent(fav, total):
    # 1st Draw: Direct fraction bana diya
    prob_A = Fraction(fav, total)
    
    # 2nd Draw: Dono mein se 1-1 kam karke fraction banaya
    prob_B_given_A = Fraction(fav - 1, total - 1)
    
    # MASTER FORMULA: P(A) * P(B|A)
    total_prob = prob_A * prob_B_given_A
    
    return {
        "P(First Draw)": prob_A,
        "P(Second Draw)": prob_B_given_A,
        "Total P(Both)": total_prob,
        "Percentage": f"{round(float(total_prob) * 100, 2)}%"
    }

# =========================================================================
# RUNNING QUESTION 9: Bag of 10 Balls (4 Red) -> Red AND Red Without Replacement
# =========================================================================
print("=" * 50)
result = short_multiplication_dependent(fav=4, total=10)

for k, v in result.items():
    print(f"{k} : {v}")
print("=" * 50)
```

```python
# =========================================================================
# MACHINE DEFINITION: Conditional Probability ki manual machine
# =========================================================================

def calculate_conditional_probability(prob_A, prob_both):
    """
    Formula: P(B | A) = P(A and B) / P(A)
    prob_A = Jo event pehle hi ho chuka hai (Given condition)
    prob_both = Dono ke ek sath hone ka chance (Intersection)
    """
    # Safety Check: Base condition ki probability 0 nahi honi chahiye (ZeroDivisionError se bachne ke liye)
    if prob_A == 0:
        return "Error: Given condition probability cannot be zero."
        
    # Safety Check: Dono ka common chance kabhi bhi akele A se bada nahi ho sakta
    if prob_both > prob_A:
        return "Error: Both-event probability cannot exceed single-event probability."

    # ASLI LOGIC: Common wale ko divide (/) karte hain pehle se huye event se
    conditional_prob = prob_both / prob_A
    percentage = conditional_prob * 100
    
    return {
        "Given Event P(A)": round(prob_A, 4),
        "Both Event P(A and B)": round(prob_both, 4),
        "Conditional Probability P(B|A)": round(conditional_prob, 4),
        "Total Percentage": f"{round(percentage, 2)}%"
    }


# =========================================================================
# ACTUAL EXECUTION: PDF ka Question 9 (GST Audit) run kar rahe hain
# =========================================================================

print("=" * 70)
# 📌 QUESTION 9 (PDF ke mutabik):
# 40% GST bills par Purchase entry galat hai (P(A) = 0.40)
# 15% bills par Purchase aur Sales dono entry galat hain (P(A and B) = 0.15)
# Agar auditor ko Purchase entry galat milti hai, toh Sales entry bhi galat hone ka kya chance hai?
print("RUNNING PDF QUESTION 9: GST Tax Audit (Conditional Probability)")
print("=" * 70)

p_purchase_wrong = 0.40
p_both_wrong = 0.15

q9_audit = calculate_conditional_probability(prob_A=p_purchase_wrong, prob_both=p_both_wrong)

# Hamara standard loop
for k, v in q9_audit.items():
    print(f"{k} : {v}")

print("=" * 70)
```
