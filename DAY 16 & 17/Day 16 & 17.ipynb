```python
import scipy.stats as stats

# Historical Factory Data Setup (Jo baseline math me tha)
mu = 100     # Historical Average (Null Hypothesis Population Mean)
sigma = 10   # Standard Deviation
alpha = 0.05 # Cut-off Border (5% Significance Level)

print("=== DAY 16 & 17: HYPOTHESIS TESTING PYTHON IMPLEMENTATION ===\n")

# Hum ek function bana lete hain taaki 116 aur 120 dono ka test ek sath ho sake
def run_hypothesis_test(x_score):
    print(f"--- TESTING PRODUCTION SCORE: {x_score} RINGS ---")
    
    # STEP 1: Z-Score Calculate Karo (Faasla Naapo)
    # Formula: Z = (x - mu) / sigma
    z_score = (x_score - mu) / sigma
    print(f"1. Calculated Z-Score: {z_score:.2f}")
    
    # STEP 2: p-value Calculate Karo (Probability / Kismat Ka Chance)
    # stats.norm.cdf(z_score) hume peeche ka area deta hai. 
    # Hume top tail ka area chahiye, isliye 1 se minus karenge (Thik Z-table math jaisa)
    p_value = 1 - stats.norm.cdf(z_score)
    print(f"2. Calculated p-Value: {p_value:.4f} (Yaani {p_value*100:.2f}%)")
    
    # STEP 3: Final Decision (Alpha se takraat)
    print("3. Verdict:")
    if p_value <= alpha:
        print(f"   => REJECT NULL HYPOTHESIS (H0) 🛑")
        print(f"   => Conclusion: Badlav Kismat ('By Chance') se nahi hua hai.")
        print(f"      Yeh sach me Nayi Machine ka REAL IMPACT hai! Machine khareed lo.\n")
    else:
        print(f"   => FAIL TO REJECT NULL HYPOTHESIS (H0) ⚠️")
        print(f"   => Conclusion: Data me mazboot saboot nahi hai.")
        print(f"      Yeh result BY CHANCE (naturally) bhi ho sakta hai. Kuch din aur test karo.\n")

# Now execution dono target inputs par:
run_hypothesis_test(116)
run_hypothesis_test(120)
```
