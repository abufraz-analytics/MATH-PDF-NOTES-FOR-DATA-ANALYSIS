```python
import random

# --- DATA SETUP ---
# Humne sabhi categories ke salaries ki alag-alag list bana li
workers = [30000, 32000, 35000, 38000, 40000, 42000, 45000] # Total 7
managers = [80000, 85000]                                  # Total 2
owners = [500000]                                          # Total 1 (Outlier)

total_population = 10
target_sample_size = 4

print("--- STEP 1 & 2: MATH SE SEATS NIKALNA ---")

# Worker ka ratio = 7/10 = 0.7. Seats = round(0.7 * 4) = 3
worker_seats = round((len(workers) / total_population) * target_sample_size)

# Manager ka ratio = 2/10 = 0.2. Seats = round(0.2 * 4) = 1
manager_seats = round((len(managers) / total_population) * target_sample_size)

# Owner ka ratio = 1/10 = 0.1. Seats = round(0.1 * 4) = 0
owner_seats = round((len(owners) / total_population) * target_sample_size)

print(f"Workers ko mili: {worker_seats} seats")
print(f"Managers ko mili: {manager_seats} seats")
print(f"Owners ko mili: {owner_seats} seats\n")


print("--- STEP 3: RANDOM SELECTION (LOTTERY) ---")

# random.sample(list, seats) khud hi randomly utha leta hai
selected_workers = random.sample(workers, worker_seats)    # 7 me se koi bhi 3 uthayega
selected_managers = random.sample(managers, manager_seats) # 2 me se koi bhi 1 uthayega
selected_owners = random.sample(owners, owner_seats)       # 1 me se 0 uthayega

# Sabko ek final sample list me mila diya (+)
final_sample = selected_workers + selected_managers + selected_owners
print(f"Hamara Final Sample Data bana: {final_sample}\n")


print("--- STEP 4: SAMPLE MEAN (AVERAGE) ---")

# Sum of salaries / Total sample size
sample_mean = sum(final_sample) / len(final_sample)
print(f"Final Sample Size: {len(final_sample)}")
print(f"Sample Mean Salary: INR {sample_mean:,.2f}")
```

```python
import pandas as pd

# 1. Raw data setup
df = pd.DataFrame({
    "role": ["Worker"]*7 + ["Manager"]*2 + ["Owner"],
    "salary": [30000, 32000, 35000, 38000, 40000, 42000, 45000, 80000, 85000, 500000]
})

# 2. THE PERFECT SHORTCUT LINE (No warning, no column loss)
sample_df = df.groupby("role").sample(frac=0.4)

# 3. Output display
print("--- Selected Sample Data ---")
print(sample_df)

print(f"\nSample Size: {len(sample_df)}")
print(f"Sample Mean Salary: INR {sample_df['salary'].mean():,.2f}")
```
