```python
data = [12, 10, 15, 20, 11]

# Step 1: Data ko Sort karna (Zaroori Step)
data.sort() # for descending order data.sort(reverse=True)
print(f"Sorted Data: {data}")
```

```python
data = [12, 10, 15, 20]

# Step 1: Data ko Sort karna (Zaroori Step)
data.sort()
n = len(data)
middle_index = n // 2  # Ye humein 2 dega

median = data[middle_index]
print(f"The Median is: {median}")
```

```python
data = [12, 10, 15, 20]

# Step 1: Sabse pehle Sort karein
data.sort() 
# Ab data ho gaya: [10, 12, 15, 20]
# Yahan beech mein do numbers hain: 12 aur 15

# Step 2: Beech ke do positions dhoondhein
n = len(data)
mid1 = n // 2      # Ye index 2 dega (yani 15)
mid2 = (n // 2) - 1 # Ye index 1 dega (yani 12)

# Step 3: Dono ka Average nikaalein
median = (data[mid1] + data[mid2]) / 2

print(f"Sorted Data: {data}")
print(f"The Median is: {median}")
```

```python
data = [18, 20, 18, 22, 18, 20, 24]

# Step 1: Har number ki ginti (Count) nikalna
ginti = {}
for i in data:
    ginti[i] = data.count(i)

# Step 2: Sabse badi ginti wala number dhoondhna
mode_value = max(ginti, key=ginti.get)

print(f"Logic se nikala gaya Mode: {mode_value}")
```

```python
import statistics
data = [10, 20, 15, 10, 25, 10, 20]

# Note: Jab aap statistics.median() use karte hain, toh ye itna smart
#hai ki chahe data Odd ho ya Even, ye apne aap sahi formula laga deta hai. 
#Aapko manually check karne ki tension nahi leni padti.
# Isme aapko data sort karne ki bhi zaroorat nahi, ye khud andar hi andar kar leta hai
median_odd = statistics.median(data)

print(f"Data: {data}")
print(f"Median (using statistics library): {median_odd}")
```

```python
import statistics

# Ring sizes sold
data = [18, 20, 18, 22, 18, 20, 24]

most_popular = statistics.mode(data)
# Multi-Mode Logic: Agar 2 items barabar matra mein bikte hain (e.g., Size 18 aur 20 dono 3-3 baar), 
# toh normal mode() error de sakta hai. 
# statistics.multimode(data) ka use karein, ye saare popular items ki list nikal ke dega.

print(f"Sabse zyada bikne wala size (Mode): {most_popular}")
# Output aayega: 18 (kyunki ye 3 baar aaya hai)
```

```python
import statistics

data = [10, 20, 30, 40]

# 1. Standard Median (Dono ka average)
# Ye 20 aur 30 ko jod kar aadha kar dega -> 25.0
print("Normal Median:", statistics.median(data))

# 2. Median Low (Chhota wala beech ka number)
# Ye 20 aur 30 mein se sirf 20 ko uthayega
print("Median Low:", statistics.median_low(data))

# 3. Median High (Bada wala beech ka number)
# Ye 20 aur 30 mein se sirf 30 ko uthayega
print("Median High:", statistics.median_high(data))
```
