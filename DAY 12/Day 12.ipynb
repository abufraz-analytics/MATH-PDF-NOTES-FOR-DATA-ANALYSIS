```python
# ==============================================================================
# DAY 12: NORMAL DISTRIBUTION - TASK 1 (QUESTION 1 ONLY)
# ==============================================================================
# AUTHOR: Abu Fraz Abbas
# 
# [INDUSTRIAL CASE STUDY - QUESTION 1]
# Scenario: Dubai Premium Gold & Diamond Park mein ek high-end precision cutting
# machine lagayi gayi hai jo custom luxury rings ke liye gold bars kaat-ti hai.
# 
# Given Parameters:
# - Mean weight (μ) = 5.00g
# - Standard Deviation (σ) = 0.02g
# 
# QUESTION 1: Machine se banni wali kitni percent rings 4.96g se 5.04g ke exact
# quality bracket mein aayengi? (Empirical Rule Verification using Manual Math)
# ==============================================================================

import math

def manual_z_score(y, mean, std_dev):
    """
    Z-SCORE TRANSFORM ENGINE
    Formula: Z = (Y - μ) / σ
    Yeh function real-world weight ko standard steps (kadamon) mein badalta hai.
    """
    z = (y - mean) / std_dev
    return z

def manual_z_table_lookup(z_score):
    """
    MANUAL Z-TABLE CDF CALCULATOR
    Printed books ke piche jo Z-table hoti hai, yeh sateek wahi value nikaalta hai.
    Yeh Minus Infinity se lekar us Z point tak ka left-tail area area deta hai.
    """
    # High-precision Error Function (erf) use karke manual cumulative area
    cumulative_area = 0.5 * (1 + math.erf(z_score / math.sqrt(2)))
    return cumulative_area

# ==============================================================================
# MAIN ENGINE EXECUTION
# ==============================================================================
if __name__ == "__main__":

    # 1. Given Business Parameters Setup
    MEAN_WEIGHT = 5.00   # μ = 5.00g
    STD_DEV = 0.02       # σ = 0.02g

    # 2. Target Quality Window Boundaries
    lower_weight = 4.96  # Bracket ki lower limit
    upper_weight = 5.04  # Bracket ki upper limit

    print("==================================================================")
    print("         GOLD & DIAMOND PARK - QUALITY BRACKET AUDIT             ")
    print("==================================================================\n")
    print(f"Machine Parameters: Mean (μ) = {MEAN_WEIGHT}g | Std Dev (σ) = {STD_DEV}g")
    print(f"Target Quality Window: {lower_weight}g se {upper_weight}g\n")

    # STEP 1: Calculate Z-Scores Manually
    # Hum pata lagayenge ki limits center se kitne standard deviation door hain
    z_lower = manual_z_score(lower_weight, MEAN_WEIGHT, STD_DEV)
    z_upper = manual_z_score(upper_weight, MEAN_WEIGHT, STD_DEV)

    print("--- STEP 1: Z-Score Steps Calculation ---")
    print(f"Lower Boundary ({lower_weight}g) ka Z-Score = {z_lower:.2f} (yani center se exact 2 kadam peeche)")
    print(f"Upper Boundary ({upper_weight}g) ka Z-Score = {z_upper:.2f} (yani center se exact 2 kadam aage)\n")

    # STEP 2: Replicate Z-Table Grid Lookups
    # Left tail se lekar un points tak ka total percentage area area nikalenge
    area_lower = manual_z_table_lookup(z_lower)
    area_upper = manual_z_table_lookup(z_upper)

    print("--- STEP 2: Manual Z-Table Area Lookup ---")
    print(f"Z-Table Area up to Lower Bound (Z = -2.00): {area_lower:.4f} (yani left side ka outer tail)")
    print(f"Z-Table Area up to Upper Bound (Z = +2.00): {area_upper:.4f} (yani center + right side tak ka area)\n")

    # STEP 3: Interval Area Deduction (Upper Area - Lower Area)
    # Beech ka sateek compliance area nikalne ke liye bade area se chhota area minus karenge
    compliance_area = area_upper - area_lower
    compliance_percentage = compliance_area * 100
    wastage_percentage = 100.0 - compliance_percentage

    print("--- STEP 3: Final Quality Compliance Metrics ---")
    print(f"Sateek Integration Window Area (P(-2.00 <= Z <= +2.00)) = {compliance_area:.4f}")
    print("------------------------------------------------------------------")
    print(f"📢 COMPLIANCE RESULT : [ {compliance_percentage:.2f}% ] rings strict quality criteria pass karengi.")
    print(f"⚠️ OUTLIERS / WASTAGE: [ {wastage_percentage:.2f}% ] rings reject hongi jinhe melt karna padega.")
    print("------------------------------------------------------------------")
```

```python
# ==============================================================================
# DAY 12: NORMAL DISTRIBUTION - SHORT LIBRARY CODE (QUESTION 1)
# ==============================================================================
# AUTHOR: Abu Fraz Abbas
# DESCRIPTION: Solving Quality Compliance using Scipy Library (Short Cut)
# ==============================================================================

import scipy.stats as stats

# 1. Given Business Parameters Setup
MEAN_WEIGHT = 5.00   # μ = 5.00g
STD_DEV = 0.02       # σ = 0.02g

# 2. Target Quality Window Boundaries
lower_weight = 4.96  # Bracket lower limit
upper_weight = 5.04  # Bracket upper limit

print("==================================================================")
# stats.norm.cdf(x, loc, scale) automatic Z-Score nikaal kar uske left-tail ka area de deta hai.
# loc = Mean (μ), scale = Standard Deviation (σ)
area_lower = stats.norm.cdf(lower_weight, loc=MEAN_WEIGHT, scale=STD_DEV)
area_upper = stats.norm.cdf(upper_weight, loc=MEAN_WEIGHT, scale=STD_DEV)

# Final Step: Beech ka area nikalne ke liye Dono Areas ko minus kiya
compliance_percentage = (area_upper - area_lower) * 100
wastage_percentage = 100.0 - compliance_percentage

# 3. Print Final Metrics
print(f"📢 COMPLIANCE RESULT : [ {compliance_percentage:.2f}% ] rings quality pass karengi.")
print(f"⚠️ OUTLIERS / WASTAGE: [ {wastage_percentage:.2f}% ] rings reject hongi.")
print("==================================================================")
```

```python
# ==============================================================================
# DAY 12: NORMAL DISTRIBUTION - PART 2 (MANUAL DATA ENGINE)
# ==============================================================================
# AUTHOR: Abu Fraz Abbas
# 
# [INDUSTRIAL CASE STUDY - QUESTION 2]
# Given Parameters:
# - Mean weight (μ) = 5.00g
# - Standard Deviation (σ) = 0.02g
# 
# QUESTION 2: Agar hume target point y = 5.02g (center se exact 1-sigma door) par
# curve ki absolute vertical height (Probability Density) nikalni ho, toh Gauss ke
# Dual Engine (Component A & Component B) ko manually kaise execute karenge?
# ==============================================================================

import math

def manual_gaussian_decomposition(y, mean, std_dev):
    """
    GAUSSIAN DUAL ENGINE DECOMPOSITION
    Formula: f(y) = [ 1 / (σ*√2π) ] * [ e^(-0.5 * ((y-μ)/σ)^2) ]
    """
    print("==================================================================")
    print("        GAUSS HEIGHT ENGINE DECOMPOSITION ENGINE (MANUAL)        ")
    print("==================================================================\n")
    print(f"Target Point (y) = {y}g | Mean (μ) = {mean}g | Std Dev (σ) = {std_dev}g\n")

    # --------------------------------------------------------------------------
    # COMPONENT A: MAXIMUM BASE HEIGHT ENGINE (NORMALIZING CONSTANT)
    # --------------------------------------------------------------------------
    # Formula: 1 / (σ * √2π)
    # Yeh component curve ke perfectly center mean (μ) par peak height fix karta hai.
    
    sqrt_two_pi = math.sqrt(2 * math.pi)       # Step 1: √2π ki value nikalna
    component_a = 1.0 / (std_dev * sqrt_two_pi) # Step 2: 1 / (σ * √2π) karna

    print("--- COMPONENT A: Maximum Base Height Engine ---")
    print(f"√2π Constant Value = {sqrt_two_pi:.4f}")
    print(f"Component A (Baseline Maximum Center Height) = {component_a:.4f}\n")


    # --------------------------------------------------------------------------
    # COMPONENT B: THE SLOPE DECAY FACTOR (% DROP)
    # --------------------------------------------------------------------------
    # Formula: e^(-0.5 * ((y-μ)/σ)^2)
    # Yeh decimal factor (0 se 1) batata hai ki center se door jaane par height kitni drop hui.
    
    # Step 1: Deviation nikalna (y - mean)
    deviation = y - mean 
    
    # Step 2: Z-Score nikalna (Z = deviation / σ)
    z_score = deviation / std_dev 
    
    # Step 3: Exponent power calculation (-0.5 * Z^2)
    exponent_power = -0.5 * (z_score ** 2)
    
    # Step 4: Exponential Base Engine (e^power)
    component_b = math.exp(exponent_power)

    print("--- COMPONENT B: The Slope Decay Factor ---")
    print(f"Real Variation Distance (y - μ) = {deviation:.3f}g")
    print(f"Z-Score Distance (Kadam) = {z_score:.2f} (yani center se exact 1 kadam door)")
    print(f"Exponent Power Multiplier = {exponent_power:.4f}")
    print(f"Component B (Percentage Height Retained) = {component_b:.4f} (yani center ka {component_b*100:.2f}% unchayi bachi)\n")


    # --------------------------------------------------------------------------
    # FINAL PROFILE HEIGHT ASSEMBLY (SYNTHESIS RESULT)
    # --------------------------------------------------------------------------
    # Final Height = Component A * Component B
    final_profile_height = component_a * component_b

    print("--- FINAL SYNTHESIS RESULT ---")
    print(f"Formula: f({y}) = Component A * Component B")
    print(f"Calculation: {component_a:.4f} * {component_b:.4f}")
    print(f"📢 ABSOLUTE VERTICAL HEIGHT f({y}) = {final_profile_height:.2f}\n")
    print("==================================================================")

    return final_profile_height

# ==============================================================================
# MAIN CORE CALL
# ==============================================================================
if __name__ == "__main__":
    
    # Case Study Values Input
    TARGET_Y = 5.02
    MEAN_WEIGHT = 5.00
    STD_DEV = 0.02
    
    # Execute the decomposition core
    manual_gaussian_decomposition(TARGET_Y, MEAN_WEIGHT, STD_DEV)
```

```python
# ==============================================================================
# DAY 12: NORMAL DISTRIBUTION - SHORT LIBRARY CODE (QUESTION 2)
# ==============================================================================
# AUTHOR: Abu Fraz Abbas
# DESCRIPTION: Absolute Coordinate Height using Scipy Library (Short Cut)
# ==============================================================================

import scipy.stats as stats

# 1. Given Business Parameters Setup
MEAN_WEIGHT = 5.00   # μ = 5.00g
STD_DEV = 0.02       # σ = 0.02g

# 2. Target Coordinate Point
target_y = 5.02      # Jis point par height nikalni hai

print("==================================================================")
# stats.norm.pdf(x, loc, scale) automatic Dual Engine ko piche se solve karke
# sateek coordinate ki absolute vertical height nikaal kar de deta hai.
# loc = Mean (μ), scale = Standard Deviation (σ)
final_height = stats.norm.pdf(target_y, loc=MEAN_WEIGHT, scale=STD_DEV)

# 3. Print Final Metric
print(f"Target Coordinate Check: y = {target_y}g")
print(f"📢 ABSOLUTE VERTICAL HEIGHT f(5.02) = {final_height:.2f}")
print("==================================================================")
```
