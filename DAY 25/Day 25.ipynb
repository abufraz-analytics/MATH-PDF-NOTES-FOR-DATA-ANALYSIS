```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Data ko Professional Matrix layout me dalna
# Rows represent Days, Columns represent Features [Weight, Complexity]
X = np.array([
    [10, 3],  # Day 1
    [30, 4],  # Day 2
    [20, 5]   # Day 3
])
Y = np.array([80000, 375000, 250000]) # Actual Prices

# 2. Model call karna aur fit (Train) kar dena
model = LinearRegression()
model.fit(X, Y)

# 3. Direct direct mathematical coefficients nikalna
m1, m2 = model.coef_ # Dono weights ek sath mil gaye
c = model.intercept_ # Fixed baseline intercept

print("--- PROFESSIONAL OPTIMIZED OUTPUT ---")
print(f"Perfect Optimized m1 (Gold Weight Slope)     : {m1:.2f}")
print(f"Perfect Optimized m2 (Complexity Slope)      : {m2:.2f}")
print(f"Perfect Optimized c (Showroom Intercept Cost): {c:.2f}")
```
