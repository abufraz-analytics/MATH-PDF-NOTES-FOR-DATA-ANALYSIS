```python
# =====================================================================
# Topic: Day 21 - Confidence Intervals & Margin of Error

print("--- STEP 1: Given Input Data (Jewelry Case Study) ---")
# PDF ke mutabiq humare paas ye data hai:
sample_size = 100         # N = 100 orders
sample_mean = 25.0        # X̄ = 25.0 grams (Sample Average)
std_deviation = 2.0       # σ = 2.0 grams (Asali Bhatkaw)
z_score = 1.96            # Z = 1.96 (95% Confidence Level ke liye fixed hai)

print("Sample Size (N) =", sample_size)
print("Sample Mean (X̄) =", sample_mean, "grams")
print("Standard Deviation (σ) =", std_deviation, "grams")
print("Fixed Z-Score for 95% =", z_score)


print("\n--- STEP 2: Standard Error (SE) Calculate Karna ---")
# Formula: SE = σ / sqrt(N)
# Hum 100 ka square root manually nikalenge jo ki 10 hota hai (10 * 10 = 100)
sqrt_n = 100 ** 0.5       # Kisi bhi number ki power 0.5 karne se uska root nikal jata hai

standard_error = std_deviation / sqrt_n   # 2.0 / 10 = 0.2
print("Square Root of N =", sqrt_n)
print("Calculated Standard Error (SE) =", standard_error)


print("\n--- STEP 3: Margin of Error (MoE) Calculate Karna ---")
# Formula: MoE = Z * SE
# Jaise aapne pakda tha, Z ko SE se multiply karke safety buffer banayenge

margin_of_error = z_score * standard_error  # 1.96 * 0.2 = 0.392
print("Calculated Margin of Error (MoE) =", margin_of_error, "grams")


print("\n--- STEP 4: Final Confidence Interval Range Build Karna ---")
# Formula: Mean me se MoE ko ghatao (Lower Limit) aur jodo (Upper Limit)

lower_limit = sample_mean - margin_of_error  # 25.0 - 0.392 = 24.608
upper_limit = sample_mean + margin_of_error  # 25.0 + 0.392 = 25.392

print("Lower Limit (Mean - MoE) =", lower_limit, "grams")
print("Upper Limit (Mean + MoE) =", upper_limit, "grams")


print("\n--- FINAL EXECUTIVE INTERPRETATION ---")
# Jo hum business report me likhenge
print(f"Hum 95% poore bharose ke sath keh sakte hain ki hamari factory ke")
print(f"sabhi orders ka asli average weight {lower_limit:.2f} grams se {upper_limit:.2f} grams ke beech hai!")
```

```python
from scipy import stats

# STEP 1: Given Data Input Karo
n, mean, sd = 100, 25.0, 2.0

# STEP 2: Single Line Me Range Nikalo (Confidence=0.95, Loc=Mean, Scale=SE)
lower_limit, upper_limit = stats.norm.interval(0.95, loc=mean, scale=sd / (n ** 0.5))

# STEP 3: Result Print Karo
print(f"95% Confidence Interval Range: [{lower_limit:.2f}, {upper_limit:.2f}] grams")
```
