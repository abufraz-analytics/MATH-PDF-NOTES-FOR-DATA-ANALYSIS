```python
import scipy.stats as stats

#  Data aur Target
coins_data = [9.8, 10.1, 1000, 10.2, 9.7, 10.1, 9.9, 10.0, 9.8, 10.1]
target_mu = 10.0

#  Sirf 1-Line me One-Sample T-Test (1samp)
t_score, p_value = stats.ttest_1samp(coins_data, popmean=target_mu)

#  Output aur Verdict
print(f"FINAL T-SCORE: {t_score:.4f} Kadam")
print(f"Verdict: {'REJECT H0' if p_value <= 0.05 else 'NULL HYPOTHESIS (H0) WON!'}")
```

```python
# =====================================================================
# DAY 18: ONE-SAMPLE T-TEST (PURE MANUAL CODE WITHOUT STATS LIBRARIES)
# Written by: Abu Fraz Abbas
# =====================================================================

#  1. Given Data & Targets
coins_data = [9.8, 10.1, 10.3, 10.2, 9.7, 10.1, 9.9, 10.0, 9.8, 10.1]
target_mu = 10.0
N = len(coins_data)  # Total coins = 10

print("--- STEP-BY-STEP MANUAL T-TEST CALCULATION ---")

#  STEP A: Sample Ka Average (Mean) Nikalna
total_sum = sum(coins_data)
sample_mean = total_sum / N
print(f"Step A: Sample Mean (X̄) = {sample_mean:.4f} grams")


#  STEP B: Sample Ki Standard Deviation (s) Nikalna
# Formula: sqrt( sum((X - X̄)^2) / (N - 1) )
squared_differences_sum = 0

for x in coins_data:
    difference = x - sample_mean
    squared_diff = difference ** 2
    squared_differences_sum += squared_diff

variance = squared_differences_sum / (N - 1)  # N - 1 is Degree of Freedom
sample_std_dev = variance ** 0.5               # Square root using power 0.5

print(f"Step B: Sum of Squared Diff = {squared_differences_sum:.4f}")
print(f"        Sample Std Deviation (s) = {sample_std_dev:.4f} grams")


#  STEP C: Standard Error (S.E.) Nikalna
# Formula: s / sqrt(N)
standard_error = sample_std_dev / (N ** 0.5)
print(f"Step C: Standard Error (S.E.) = {standard_error:.5f} grams")


#  STEP D: Final T-Score (The Division / Guna Check)
# Formula: (X̄ - mu) / S.E.
mean_difference = sample_mean - target_mu
t_score = mean_difference / standard_error

print(f"Step D: Mean Difference = {mean_difference:.4f} grams")
print(f"=================================================")
print(f"FINAL T-SCORE (Kitna Guna Jhatka) = {t_score:.4f} Kadam")
print(f"=================================================")


#  INDUSTRIAL VERDICT (Faisla)
print("\n--- FINAL INDUSTRIAL VERDICT ---")
# Agar T-Score -2 aur +2 ke beech me hai, to variation normal hai
if abs(t_score) <= 2:
    print("Verdict: Null Hypothesis (H0) Won!")
    print("Machine ekdum perfect chal rahi hai, jo chhota farq hai woh bas kismat ka jhatka hai.")
else:
    print("Verdict: Reject H0! Machine me kharabi hai!")
    print("Difference kismat ke jhatke se 2 guna se zyada bada hai, machine check karo!")
```
