```python
# =====================================================================
# DAY 11: DISCRETE RANDOM VARIABLE (PMF) - MANUAL ENGINE
# =====================================================================

# Step 1: Input Data Lists (Hamara manual sample space aur initial weights)
x_values = [0, 1, 2, 3, 4]  # X = Number of luxury jewelry orders per hour

# P(X = x) ki raw values jisme k = 0.08 humne maths se nikala tha
# P(X=0)=0.40, P(X=1)=0.25, P(X=2)=0.11, P(X=3)=2k (0.16), P(X=4)=k (0.08)
pmf_probabilities = [0.40, 0.25, 0.11, 0.16, 0.08]

# ---------------------------------------------------------------------
# CONDITION 1 CHECKER: Verify validation axioms via manual loop
# ---------------------------------------------------------------------
print("--- STAGE 1: MANUAL AXIOM VALIDATION CHECK ---")

total_pmf_sum = 0.0
is_valid_pmf = True

# Manual loop chalakar har ek value ko check aur sum kar rahe hain
for i in range(len(pmf_probabilities)):
    current_prob = pmf_probabilities[i]
    
    # Check A: Koi bhi probability negative (< 0) nahi honi chahiye
    if current_prob < 0:
        print(f"-> ALERT: Negative probability found at index {i}!")
        is_valid_pmf = False
        
    total_pmf_sum += current_prob

print(f"-> Calculated Total Sum of PMF Box = {total_pmf_sum:.2f}")

# Check B: Total sum exact 1.0 (100%) hona chahiye
if total_pmf_sum == 1.0 and is_valid_pmf:
    print("-> RESULT: PMF validation successful! All axioms satisfied.")
else:
    print("-> RESULT: PMF validation failed!")

print("=" * 60)

# ---------------------------------------------------------------------
# TARGET CALCULATION: Finding P(X >= 3) via explicit looping
# ---------------------------------------------------------------------
print("--- STAGE 2: TARGET PROBABILITY EXECUTION [P(X >= 3)] ---")

target_probability = 0.0

# Table ke har ek row par manual evaluation matrix chala rahe hain
for i in range(len(x_values)):
    current_x = x_values[i]
    
    # Condition Trigger: Jab hamara sateek number 'x' >= 3 hoga, tabhi sum karenge
    if current_x >= 3:
        target_probability += pmf_probabilities[i]
        print(f"-> Target Condition Matched! Included: P(X = {current_x}) = {pmf_probabilities[i]:.2f}")

# Final Result Extraction
final_percentage = target_probability * 100
print("-" * 60)
print(f"FINAL BUSINESS REPORT OUTCOME:")
print(f"The exact probability P(X >= 3) is: {target_probability:.2f} (or {final_percentage:.2f}%)")
print("=" * 60)
```

```python
import numpy as np

# Step 1: Data ko NumPy Arrays mein set kiya
x_values = np.array([0, 1, 2, 3, 4])
pmf_probabilities = np.array([0.40, 0.25, 0.11, 0.16, 0.08])

# ---------------------------------------------------------------------
# SHORTCUT CODE: Single Line Condition Filter & Sum
# ---------------------------------------------------------------------

# 1. Condition applied directly on array: Isse backend par mask (True/False) banega
# 2. np.sum() sirf unhi probabilities ko jodega jahan condition True hai
target_probability = np.sum(pmf_probabilities[x_values >= 3])

# Final Output Report
print(f"P(X >= 3) Short Output: {target_probability * 100:.2f}%")
```

```python
# =====================================================================
# DAY 11: CONTINUOUS RANDOM VARIABLE (PDF) - MANUAL ENGINE
# =====================================================================

# Step 1: Core Parameters Setting
# operational_min = 0.0, operational_max = 2.0 (Hamari universe boundary)
# Target window: Humse poocha hai 1 se 2 ghante ke beech ka chance

c_constant = 0.5  # Jo humne maths se nikala tha f(y) = 0.5 * y
dy = 0.00001      # Strikingly small time step (Width of one tiny rectangle)

# ---------------------------------------------------------------------
# CONDITION 2 CHECKER: Verify if Total Area under the curve equals 1.0
# ---------------------------------------------------------------------
print("--- STAGE 1: MANUAL CONTINUOUS TOTAL AREA CHECK ---")

total_area = 0.0
current_y = 0.0   # Operational Minimum se shuru karenge
is_valid_pdf = True

# Manual loop: Range Min (0.0) se lekar Max (2.0) tak chalega dy ke steps mein
while current_y < 2.0:
    # Calculate Height of current rectangle using the formula f(y) = c * y
    height = c_constant * current_y
    
    # Check A: Density height kabhi minus mein nahi jani chahiye
    if height < 0:
        is_valid_pdf = False
        
    # Area of tiny rectangle = Height * Width
    tiny_area = height * dy
    total_area += tiny_area
    
    # Move tracker forward by step size dy
    current_y += dy

print(f"-> Calculated Total Area under the curve = {total_area:.4f}")

if abs(total_area - 1.0) < 0.001 and is_valid_pdf:
    print("-> RESULT: PDF validation successful! Total Area is exactly 1.")
else:
    print("-> RESULT: PDF validation failed!")

print("=" * 60)

# ---------------------------------------------------------------------
# TARGET CALCULATION: Finding P(1 <= Y <= 2) via explicit looping
# ---------------------------------------------------------------------
print("--- STAGE 2: TARGET WINDOW RISK EXECUTION [P(1 <= Y <= 2)] ---")

target_range_area = 0.0
current_y = 1.0   # Sawaal ki Lower Limit (1.0) se shuru karenge

# Manual loop: Target window ke lower limit (1.0) se upper limit (2.0) tak chalega
while current_y < 2.0:
    height = c_constant * current_y
    tiny_area = height * dy
    target_range_area += tiny_area
    
    current_y += dy

# Final Result Extraction
final_percentage = target_range_area * 100
print("-" * 60)
print(f"FINAL BUSINESS REPORT OUTCOME:")
print(f"The exact probability P(1 <= Y <= 2) is: {target_range_area:.4f} (or {final_percentage:.2f}%)")
print("=" * 60)
```

```python
from scipy.integrate import quad

# Step 1: Hamare machine-generated formula ka ek chota function banaya
# f(y) = 0.5 * y
def pdf_formula(y):
    return 0.5 * y

# ---------------------------------------------------------------------
# SHORTCUT ENGINE: quad(function, lower_limit, upper_limit)
# ---------------------------------------------------------------------

# 1. Total Area Check (Boundary: 0 se 2 ghante)
# quad hamari math ki tarah calculus integration chalata hai backend par
total_area, error = quad(pdf_formula, 0, 2)
print(f"Short Code -> Total Area: {total_area:.2f}")

# 2. Target Window Probability (Sawaal: 1 se 2 ghante ke beech ka chance)
target_probability, error = quad(pdf_formula, 1, 2)

print(f"Short Code -> P(1 <= Y <= 2): {target_probability * 100:.2f}%")
```
