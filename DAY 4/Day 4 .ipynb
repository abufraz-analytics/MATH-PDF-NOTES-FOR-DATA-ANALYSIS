```python
# Karigar B ka gold chain production data (in grams)
data = [5, 25, 12, 18, 9]

# Step 1: Built-in functions ka use karke direct Max aur Min nikalna
sabse_bada = max(data)      # direct 25 dhoondh lega
sabse_chhota = min(data)    # direct 5 dhoondh lega

# Step 2: Range nikalna
range_value = sabse_bada - sabse_chhota

# Step 3: Result print karna
print(f"Sabse Badi Value (Max): {sabse_bada}")
print(f"Sabse Chhoti Value (Min): {sabse_chhota}")
print(f"Data ki Range hai: {range_value}")
```

```python
# Gold chain production data (in grams)
data = [10, 20, 15]

# STEP 1: Mean (Average) nikalna
total_sum = sum(data)
total_elements = len(data)
mean = total_sum / total_elements

# STEP 2: Har element ka Mean se farq (deviation) nikal kar uska Square karna
squares_list = []
for x in data:
    farq = x - mean          # Mean se dhoori nikali
    square_farq = farq ** 2  # Minus sign khatam karne ke liye square kiya
    squares_list.append(square_farq)

# STEP 3: Un saare squared values ka average (Variance) nikalna
variance = sum(squares_list) / len(squares_list)

# STEP 4: Results print karna
print(f"1. Data ka Mean hai: {mean}")
print(f"2. Squared Deviations ki list: {squares_list}")
print(f"3. Final Manual Variance hai: {variance:.2f}")
```

```python
import statistics

# Karigar ka production data (sirf 3 elements ka sample/population)
data = [10, 20, 15]

# -------------------------------------------------------------------------
# TYPE 1: POPULATION VARIANCE (pvariance)
# -------------------------------------------------------------------------
# English: Used when you have 100% of the data. It divides the total sum of squares by 'n'.
# Hindi: Jab hamare paas poora ka poora (100%) data maujood ho. Ye total squares ko 'n' se divide karta hai.
pop_variance = statistics.pvariance(data)


# -------------------------------------------------------------------------
# TYPE 2: SAMPLE VARIANCE (variance)
# -------------------------------------------------------------------------
# English: Used when data is just a small sample of a bigger group. It divides by 'n - 1' to keep a safe margin.
# Hindi: Jab data kisi bade group ka ek chhota sa tukda (sample) ho. Ye safe margin rakhne ke liye 'n - 1' se divide karta hai.
samp_variance = statistics.variance(data)


# -------------------------------------------------------------------------
# RESULTS PRINT KARNA
# -------------------------------------------------------------------------
print(f"1. Population Variance (Divided by 3): {pop_variance:.2f}")
print(f"2. Sample Variance (Divided by 3 - 1 = 2): {samp_variance:.2f}")

# -------------------------------------------------------------------------
# MANUAL MATH BACKEND (Parde ke peeche kya hua):
# -------------------------------------------------------------------------
# Mean = (10 + 20 + 15) / 3 = 15
# Squares = (10-15)² + (20-15)² + (15-15)² = 25 + 25 + 0 = 50
#
# Population Formula = 50 / 3 = 16.67 (strict and exact)
# Sample Formula     = 50 / 2 = 25.00 (extra safety margin)
```

```python
import math

# Gold chain production data (in grams)
data = [10, 20, 15]

# STEP 1: Mean (Average) nikalna
mean = sum(data) / len(data)

# STEP 2: Variance wala loop chalana (Population Variance)
squares_list = []
for x in data:
    farq = x - mean
    squares_list.append(farq ** 2)

variance = sum(squares_list) / len(data)  # Answer aaya tha 16.67

# STEP 3: SD nikalna (Variance ka Square Root)
# English: Taking the square root to bring the unit back from grams² to grams.
# Hindi: Variance ka square root lena taaki unit grams² se wapas normal grams mein aa jaye.
std_deviation = math.sqrt(variance)

# STEP 4: Results print karna
print(f"1. Manual Population Variance: {variance:.2f}")
print(f"2. Final Manual Standard Deviation (SD): {std_deviation:.2f}")
```

```python
import statistics

# Gold chain production data (in grams)
data = [10, 20, 15]

# Step 1: statistics library ke pstdev function ka use karna (Population SD)
# English: This directly takes the square root of the population variance (divided by n).
# Hindi: Ye direct population variance (jo n se divide hua tha) uska square root nikalta hai.
pop_sd = statistics.pstdev(data)

# Step 2: statistics library ke stdev function ka use karna (Sample SD)
# English: This takes the square root of the sample variance (divided by n-1).
# Hindi: Ye sample variance (jo n-1 se divide hua tha) uska square root nikalta hai.
samp_sd = statistics.stdev(data)

# Step 3: Results print karna
print(f"1. Library Population SD (Strict Risk): {pop_sd:.2f}")
print(f"2. Library Sample SD (With Safety Margin): {samp_sd:.2f}")
```

```python
import statistics

# Dataset 1: Gold Prices (Bada Average, Chhota Relative Bhatkaav)
gold_prices = [4950, 5050, 5000]

# Dataset 2: Silver Prices (Chhota Average, Bada Relative Bhatkaav)
silver_prices = [90, 110, 100]

# -------------------------------------------------------------------------
# STEP 1: Gold ka Mean, SD aur CV nikalna
# -------------------------------------------------------------------------
gold_mean = statistics.mean(gold_prices)
gold_sd = statistics.pstdev(gold_prices) # Population SD use kar rahe hain
gold_cv = (gold_sd / gold_mean) * 100

# -------------------------------------------------------------------------
# STEP 2: Silver ka Mean, SD aur CV nikalna
# -------------------------------------------------------------------------
silver_mean = statistics.mean(silver_prices)
silver_sd = statistics.pstdev(silver_prices)
silver_cv = (silver_sd / silver_mean) * 100

# -------------------------------------------------------------------------
# STEP 3: Results aur Sacha Muqabla Print Karna
# -------------------------------------------------------------------------
print("--- GOLD MARKET ANALYSIS ---")
print(f"Gold Mean: {gold_mean}")
print(f"Gold Standard Deviation (SD): {gold_sd:.2f}")
print(f"Gold CV (Risk %): {gold_cv:.2f}%\n")

print("--- SILVER MARKET ANALYSIS ---")
print(f"Silver Mean: {silver_mean}")
print(f"Silver Standard Deviation (SD): {silver_sd:.2f}")
print(f"Silver CV (Risk %): {silver_cv:.2f}%\n")

# Step 4: Decision Logic
if silver_cv > gold_cv:
    print("Conclusion: Silver market zyada volatile (risky) hai!")
else:
    print("Conclusion: Gold market zyada volatile (risky) hai!")
```
